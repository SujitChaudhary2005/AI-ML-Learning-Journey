### Interactive Plots and Widgets 

### What are Interactive Plots and Widgets?

Interactive plots allow us to go **beyond static charts** by adding elements like zooming, panning, sliders, buttons, and dropdowns. In Matplotlib, we can enable interactivity using the **`%matplotlib notebook` magic command** inside Jupyter Notebook. This changes the backend from static (`inline`) to interactive, meaning plots can be rotated, zoomed, and dynamically updated without re-rendering.

Widgets in Matplotlib (and `ipywidgets`) make it possible to **manipulate plots in real time**. For example, we can use a **slider to adjust parameters**, a **button to reset data**, or a **dropdown to change variables** being plotted. This is extremely powerful for data exploration because it allows us to test hypotheses and see results instantly.

In **AI/ML and data science**, interactive plots and widgets are useful for:

- Exploring hyperparameters (e.g., learning rate vs. accuracy).
- Visualizing how model performance changes with different datasets.
- Inspecting subsets of large datasets dynamically.
- Making dashboards for presentations or teaching.

To enable interactivity:

```python
import matplotlib.pyplot as plt
```

After that, we can combine **Matplotlib + ipywidgets** to create sliders, checkboxes, and interactive controls.

### Common Interactive Plot Examples

1. Simple Interactive Line Plot

````python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.figure()
plt.plot(x, y)
plt.title("Interactive Line Plot - Pan & Zoom Enabled")
plt.xlabel("X")
plt.ylabel("sin(x)")
plt.show()
````

2. Interactive Slider with ipywidgets

````python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact

x = np.linspace(0, 10, 200)

def plot_sine(freq=1.0):
    y = np.sin(freq * x)
    plt.figure(figsize=(8,4))
    plt.plot(x, y)
    plt.title(f"Sine Wave with Frequency {freq}")
    plt.show()

interact(plot_sine, freq=(0.1, 5.0, 0.1))
````

3. Dropdown for Plot Selection

````python
from ipywidgets import interact

def choose_plot(plot_type="line"):
    if plot_type == "line":
        plt.plot(x, np.sin(x))
        plt.title("Line Plot")
    elif plot_type == "scatter":
        plt.scatter(x, np.sin(x))
        plt.title("Scatter Plot")
    plt.show()

interact(choose_plot, plot_type=["line", "scatter"])
````

4. Interactive Random Data Generator with Button

````python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import widgets

button = widgets.Button(description="Generate New Data")

def on_button_click(b):
    plt.figure()
    plt.plot(np.random.rand(10), marker="o")
    plt.title("New Random Data")
    plt.show()

button.on_click(on_button_click)
button
````

### Exercises

Q1. Create an interactive sine and cosine plot with a frequency slider.

````python
def plot_sine_cos(freq=1.0):
    y1 = np.sin(freq * x)
    y2 = np.cos(freq * x)
    plt.figure(figsize=(8,4))
    plt.plot(x, y1, label="sin(x)")
    plt.plot(x, y2, label="cos(x)")
    plt.legend()
    plt.show()

interact(plot_sine_cos, freq=(0.5, 5.0, 0.5))
````

Q2. Make a dropdown to switch between plotting **Sales** and **Profit** from the Superstore dataset.

````python
import pandas as pd
df = pd.read_csv("data/Superstore.csv", encoding='latin-1')

def plot_column(col="Sales"):
    plt.figure(figsize=(8,4))
    plt.plot(df[col].head(100))
    plt.title(f"{col} (First 100 Rows)")
    plt.show()

interact(plot_column, col=["Sales", "Profit"])
````

Q3. Create an interactive scatter plot of Sales vs Profit, where marker size is controlled by a slider.

````python
def scatter_plot(size=50):
    plt.figure(figsize=(8,6))
    plt.scatter(df["Sales"].head(200), df["Profit"].head(200), s=size, alpha=0.6)
    plt.title("Sales vs Profit (Interactive Size)")
    plt.xlabel("Sales")
    plt.ylabel("Profit")
    plt.show()

interact(scatter_plot, size=(10, 200, 10))
````

Q4. Add a button to generate a new random bar chart each time it’s clicked.

````python
button = widgets.Button(description="New Bar Chart")

def plot_bars(b):
    plt.figure()
    plt.bar(range(5), np.random.randint(1, 20, 5))
    plt.title("Random Bar Chart")
    plt.show()

button.on_click(plot_bars)
button
````

Q5. Build an interactive plot to show a polynomial curve where the **degree** is controlled by a slider.

````python
def polynomial_plot(degree=2):
    y = x**degree
    plt.figure(figsize=(8,4))
    plt.plot(x, y, label=f"x^{degree}")
    plt.legend()
    plt.show()

interact(polynomial_plot, degree=(1, 6, 1))
````

### Summary

Interactive plots and widgets transform Matplotlib from a **static visualization library** into a **dynamic exploration tool**. By enabling `%matplotlib notebook` and combining it with `ipywidgets`, we can pan, zoom, and update plots in real time.

Key takeaways:

- **Interactive line/scatter plots** allow zooming and panning inside Jupyter.
- **Sliders and dropdowns** let us manipulate parameters like frequency, dataset choice, or marker size.
- **Buttons** can trigger actions like generating new data on demand.
- These features are extremely valuable in **data exploration, AI model tuning, and presentations**.

In practice, interactive visualization helps:

- Experiment with **hyperparameters dynamically** (like degree of a polynomial).
- Quickly explore **different dataset features** without rewriting code.
- Build **mini dashboards** for teaching, analysis, or reporting.

Thus, interactive plots are not just about making charts look fancier—they **improve productivity, intuition, and communication of insights** in data science and AI/ML.
