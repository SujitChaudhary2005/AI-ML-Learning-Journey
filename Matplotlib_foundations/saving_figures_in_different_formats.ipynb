# Saving Figures in Different Formats

When creating plots in Matplotlib, it’s often necessary to save them for reports, presentations, or publications. Instead of taking screenshots, Matplotlib provides the `savefig()` function, which allows us to export plots in multiple formats with control over resolution, background, and layout. This ensures professional-quality figures that are ready for sharing or publishing.

### What is Saving Figures in Matplotlib?

The function `plt.savefig()` is used to save the current figure. We can choose the **file format**, **DPI (resolution)**, and **background style**.

**Basic Save**

```python
plt.savefig("figure.png")
```

**Save with DPI (Resolution)**

```python
plt.savefig("figure.png", dpi=300)   # Higher DPI = sharper image
```

**Save Without Extra White Space**

```python
plt.savefig("figure.png", bbox_inches="tight")
```

**Save with Transparent Background**

```python
plt.savefig("figure.png", transparent=True)
```

### Common File Formats

- **PNG, JPG** → Best for web, slides, and general use.
- **PDF, SVG, EPS** → Vector graphics, infinite scalability, perfect for printing and publications.
- **TIFF** → Very high-quality format used in printing.

### Example

```python
import matplotlib.pyplot as plt
import pandas as pd

# Sample dataset
df = pd.read_csv("data/Superstore.csv", encoding='latin-1')

# Plot Sales
plt.plot(df['Sales'][:100], color="blue", label="Sales")
plt.title("Sales (First 100 Rows)")
plt.xlabel("Order Index")
plt.ylabel("Sales ($)")
plt.legend()
plt.grid(True, linestyle="--", alpha=0.6)

# Save in multiple formats
plt.savefig("save_plots/sales_plot.png", dpi=300)          # High-quality PNG
plt.savefig("save_plots/sales_plot.pdf")                   # Vector format (PDF)
plt.savefig("save_plots/sales_plot.svg", bbox_inches="tight")  # SVG with tight layout
plt.show()
```

This example shows how the same figure can be saved in PNG (for slides), PDF (for printing), and SVG (for editing in design software).

### Exercises

Q1. Plot **Sales for the first 100 rows** and save it as a high-resolution PNG with `dpi=300`.

```python
plt.plot(df["Sales"].head(100), color="blue")
plt.title("Sales (First 100 Rows)")
plt.xlabel("Order Index")
plt.ylabel("Sales ($)")
plt.grid(True)
plt.savefig("save_plots/sales_highres.png", dpi=300)
plt.show()
```

Q2. Plot **Profit for the first 50 rows** and save it as a PDF file.

```python
plt.plot(df["Profit"].head(50), color="red")
plt.title("Profit (First 50 Rows)")
plt.xlabel("Order Index")
plt.ylabel("Profit")
plt.grid(True)
plt.savefig("save_plots/profit_plot.pdf")
plt.show()
```

Q3. Plot **Quantity (first 30 rows)** and save it with a transparent background.

```python
plt.bar(range(30), df["Quantity"].head(30), color="green")
plt.title("Quantity (First 30 Rows)")
plt.xlabel("Index")
plt.ylabel("Quantity")
plt.savefig("save_plots/quantity_transparent.png", transparent=True, dpi=200)
plt.show()
```

Q4. Plot **Discount (first 40 rows)** and save in both **PNG and SVG** formats.

```python
plt.plot(df["Discount"].head(40), color="purple", marker="o")
plt.title("Discount (First 40 Rows)")
plt.xlabel("Order Index")
plt.ylabel("Discount")
plt.grid(True)

plt.savefig("save_plots/discount_plot.png", dpi=200)
plt.savefig("save_plots/discount_plot.svg")
plt.show()
```

Q5. Create a **scatterplot of Sales vs Profit** and save it as a high-resolution TIFF file.

```python
plt.scatter(df["Sales"].head(50), df["Profit"].head(50), color="orange", alpha=0.6)
plt.title("Sales vs Profit Scatterplot")
plt.xlabel("Sales ($)")
plt.ylabel("Profit ($)")
plt.grid(True)
plt.savefig("save_plots/sales_profit.tiff", dpi=300)
plt.show()
```

### Summary

Saving figures in Matplotlib is an essential step in data visualization, especially when preparing reports, academic papers, or presentations. Using the `savefig()` function, we can export plots into different formats such as **PNG, JPG, PDF, SVG, and TIFF**, each serving different purposes.

- **PNG/JPG** are best for everyday use, presentations, and web.
- **PDF/SVG/EPS** provide **infinite scalability** and are ideal for publishing or design.
- **TIFF** ensures the highest quality for printing.
- Options like `dpi`, `bbox_inches`, and `transparent` give us precise control over resolution, layout, and background.

By mastering saving figures, we ensure our visualizations maintain **professional quality** across any medium.
