# Figure and Axes Objects: Understanding the Anatomy of a Plot

### What are figure and axes object?

When we plot in Matplotlib, we often use `plt.plot()` directly. But behind the scenes, Matplotlib uses a **Figure** (the entire canvas) and **Axes** (the actual plot area where data is drawn).

To create **professional and complex visualizations**, it’s very important to understand this **anatomy of a plot**.

Think of it like this:

- **Figure →** A paper sheet (canvas).
- **Axes →** A rectangular drawing area inside the sheet where data is plotted.
- **Axis →** The x-axis and y-axis inside the Axes.

This **Object-Oriented (OO) approach** in Matplotlib gives us much finer control than just using `plt.plot()`.

### Key Components

1. **Figure (`plt.figure` or `fig`)**
- The **whole canvas** on which one or more plots can be drawn.
- We can have multiple Axes (subplots) inside one Figure.

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(8, 6))  # Creates a blank figure
```

1. **Axes (`fig.add_axes()` or `fig.add_subplot()` or `plt.subplots()`)**
- The **actual plot area** where data appears.
- Each Axes contains:
    - **x-axis and y-axis**
    - **labels, titles, ticks**
    - The plotted data itself

```python
fig, ax = plt.subplots()  # Most common way
ax.plot([1, 2, 3], [4, 5, 6])
```

1. **Axis** (x-axis, y-axis)
    - Responsible for controlling scale, ticks, and tick labels.
    - Example: `ax.set_xlabel("X-axis")`, `ax.set_ylabel("Y-axis")`.
2. **Other elements of a plot:**
    - **Title** → `ax.set_title("My Plot")`
    - **Legend** → `ax.legend()`
    - **Grid** → `ax.grid(True)`

### Example

```python
# Create Figure and Axes
fig, ax = plt.subplots(figsize=(7, 5))

# Plot data on Axes
ax.plot([10, 20, 30, 40], [1, 4, 9, 16], label="Squared")

# Customizations
ax.set_title("Figure and Axes Example")
ax.set_xlabel("X Values")
ax.set_ylabel("Y Values")
ax.legend()
ax.grid(True)

plt.show()
```

### Exercises

Loading Dataset

```python
import pandas as pd
import numpy as np

df = pd.read_csv("data/Superstore.csv", encoding='latin-1')
```

Q1. Create a blank `Figure` object of size (6,4).

```python
fig = plt.figure(figsize=(6,4))
plt.show()
```

Q2. Add an `Axes` to it manually using `fig.add_axes([0.1,0.1,0.8,0.8])` and plot some random data.

```python
fig = plt.figure(figsize=(6,4))
ax = fig.add_axes([0.1,0.1,0.8,0.8])  # [left, bottom, width, height]
x = np.arange(10)
y = np.random.randint(1,20,10)
ax.plot(x, y, marker="o", color="blue")
ax.set_title("Manual Axes with Random Data")
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
plt.show()
```

Q3. Use `plt.subplots(2,2)` to create a **2x2 grid of plots** and fill each subplot with different charts (line, bar, scatter, histogram).

```python
fig, axes = plt.subplots(2,2, figsize=(10,8))

# Line Plot
axes[0,0].plot(df['Sales'][:50], color="blue")
axes[0,0].set_title("Line Plot of Sales")

# Bar Plot
axes[0,1].bar(np.arange(20), df['Profit'][:20], color="orange")
axes[0,1].set_title("Bar Plot of Profit")

# Scatter Plot
axes[1,0].scatter(df['Sales'][:50], df['Profit'][:50], color="green")
axes[1,0].set_title("Scatter: Sales vs Profit")

# Histogram
axes[1,1].hist(df['Sales'], bins=20, color="purple", alpha=0.7)
axes[1,1].set_title("Histogram of Sales")

# Adjust layout
plt.tight_layout()
plt.show()
```

Q4. Customize titles, labels, and grid for each Axes separately.

```python
fig, axes = plt.subplots(2,2, figsize=(10,8))

# Line Plot
axes[0,0].plot(df['Sales'][:50], color="blue")
axes[0,0].set_title("Line Plot of Sales")
axes[0,0].set_xlabel("Index")
axes[0,0].set_ylabel("Sales")
axes[0,0].grid(True)

# Bar Plot
axes[0,1].bar(np.arange(20), df['Profit'][:20], color="orange")
axes[0,1].set_title("Bar Plot of Profit")
axes[0,1].set_xlabel("Index")
axes[0,1].set_ylabel("Profit")
axes[0,1].grid(True)

# Scatter Plot
axes[1,0].scatter(df['Sales'][:50], df['Profit'][:50], color="green")
axes[1,0].set_title("Scatter: Sales vs Profit")
axes[1,0].set_xlabel("Sales")
axes[1,0].set_ylabel("Profit")
axes[1,0].grid(True)

# Histogram
axes[1,1].hist(df['Sales'], bins=20, color="purple", alpha=0.7)
axes[1,1].set_title("Histogram of Sales")
axes[1,1].set_xlabel("Sales")
axes[1,1].set_ylabel("Frequency")
axes[1,1].grid(True)

plt.tight_layout()
plt.show()
```

Q5. Try creating two Axes in the same Figure with **different scales** (Sales vs Profit).

```python
fig, ax1 = plt.subplots(figsize=(7,5))

# First axis - Sales
ax1.plot(df['Sales'][:50], color="blue", label="Sales")
ax1.set_xlabel("Index")
ax1.set_ylabel("Sales", color="blue")
ax1.tick_params(axis="y", labelcolor="blue")

# Second axis (twin) - Profit
ax2 = ax1.twinx()
ax2.plot(df['Profit'][:50], color="red", label="Profit")
ax2.set_ylabel("Profit", color="red")
ax2.tick_params(axis="y", labelcolor="red")

# Title
plt.title("Sales vs Profit with Twin Axes")
fig.tight_layout()
plt.show()
```

### Summary

The **Figure and Axes objects** form the backbone of all plots in Matplotlib. A **Figure** is essentially the container or canvas that holds everything — it can include multiple **Axes** objects, legends, titles, and annotations. The **Axes** object is where the actual plotting happens. Each Axes contains an x-axis and y-axis, tick marks, grid lines, and labels. Without Axes, we cannot visualize data.

When we use `plt.plot()`, Matplotlib automatically creates a default Figure and Axes for us. But in professional data visualization, especially when building dashboards or working with multiple subplots, it’s better to explicitly create Figures and Axes using `plt.subplots()`. This approach gives full control — we can adjust size, style, spacing, and position of each plot.

For instance, a Figure might represent a single report page, while Axes represent different charts on that page. Each Axes can be customized independently: different labels, scales, or even different chart types (line, bar, scatter, etc.).

Understanding this anatomy makes plotting flexible. It also allows us to build **presentation-ready, publication-quality visualizations**. Without this foundation, working with legends, spines, twin axes, or multiple subplots becomes difficult.

So, mastering the **Figure–Axes hierarchy** is essential for moving from basic plotting to advanced visualization workflows.
