# Axis Labels, Titles, and Legends

When creating visualizations, the raw chart alone is rarely enough to communicate the message clearly. To make plots meaningful and presentation-ready, we must add **axis labels**, a **title**, and a **legend**. These elements turn a simple visualization into an informative one.

### Axis Labels

Axis labels describe what each axis represents. In Matplotlib:

- `plt.xlabel("X-axis Label")` → adds a label to the x-axis.
- `plt.ylabel("Y-axis Label")` → adds a label to the y-axis.

In the Object-Oriented (OO) approach:

- `ax.set_xlabel("X-axis Label")`
- `ax.set_ylabel("Y-axis Label")`

Labels help the audience quickly understand the variables being visualized.

### Titles

A plot should always have a clear title. In Matplotlib:

- `plt.title("My Plot Title")`
- In OO: `ax.set_title("My Plot Title")`

We can also customize titles:

```python
import matplotlib.pyplot as plt

plt.title("Sales Trend", fontsize=14, color="blue", loc="left")
```

Here `fontsize`, `color`, and `loc` (`left`, `center`, `right`) customize the look and placement.

### Legend

A legend explains what each line, marker, or color in the plot represents.

- Using Pyplot API:

```python
import pandas as pd

# Load dataset
df = pd.read_csv("data/Superstore.csv", encoding='latin-1')

plt.plot(df['Sales'][:50], label="Sales")
plt.plot(df['Profit'][:50], label="Profit")
plt.legend()
```

- In OO API:

```python
fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(df['Sales'][:50], label="Sales")
ax.plot(df['Profit'][:50], label="Profit")
ax.legend(loc="upper left")
```

We can control the legend’s location with `loc` (`upper right`, `lower left`, etc.) or by coordinates.

### Example

```python
# Plot Sales and Profit
plt.figure(figsize=(8,5))
plt.plot(df['Sales'][:50], label="Sales", color="blue", marker="o")
plt.plot(df['Profit'][:50], label="Profit", color="red", linestyle="--", marker="x")

# Add labels, title, and legend
plt.xlabel("Order Index")
plt.ylabel("Amount ($)")
plt.title("Sales and Profit Trends (First 50 Orders)", fontsize=14, color="darkgreen")
plt.legend(loc="upper right")
plt.grid(True)

plt.show()
```

### Exercises

Q1. Plot Sales for the first 100 rows, add axis labels (“Order Index” and “Sales ($)”), and give the chart a title.

```python
plt.figure(figsize=(8,4))
plt.plot(df['Sales'][:100], color='blue')
plt.xlabel("Order Index")
plt.ylabel("Sales ($)")
plt.title("Sales for First 100 Orders")
plt.grid(True)
plt.show()
```

Q2. Plot Profit vs Sales, add both axis labels, and title it “Profit vs Sales Scatterplot.”

```python
plt.figure(figsize=(6,4))
plt.scatter(df['Sales'][:100], df['Profit'][:100], color='green')
plt.xlabel("Sales ($)")
plt.ylabel("Profit ($)")
plt.title("Profit vs Sales Scatterplot")
plt.grid(True)
plt.show()
```

Q3. Plot Sales and Quantity for the first 50 rows on the same graph, add a legend to differentiate them.

```python
plt.figure(figsize=(8,4))
plt.plot(df['Sales'][:50], label="Sales", color="blue", linestyle="-", marker="o")
plt.plot(df['Quantity'][:50], label="Quantity", color="red", linestyle="--", marker="x")
plt.xlabel("Order Index")
plt.ylabel("Values")
plt.title("Sales and Quantity for First 50 Orders")
plt.legend()
plt.grid(True)
plt.show()
```

Q4. Customize a plot with a large title in blue, axis labels in bold, and a legend in the bottom-right corner.

```python
plt.figure(figsize=(8,4))
plt.plot(df['Profit'][:50], label="Profit", color="purple")
plt.plot(df['Discount'][:50], label="Discount", color="orange")
plt.xlabel("Order Index", fontweight="bold")
plt.ylabel("Values", fontweight="bold")
plt.title("Profit vs Discount (First 50 Orders)", fontsize=14, color="blue", fontweight="bold")
plt.legend(loc="lower right")
plt.grid(True)
plt.show()
```

Q5. Plot all four variables (Sales, Profit, Quantity, Discount) for the first 30 rows with a legend and descriptive title.

```python
plt.figure(figsize=(10,6))
plt.plot(df['Sales'][:30], label="Sales", marker="o")
plt.plot(df['Profit'][:30], label="Profit", marker="x")
plt.plot(df['Quantity'][:30], label="Quantity", marker="s")
plt.plot(df['Discount'][:30], label="Discount", marker="^")
plt.xlabel("Order Index")
plt.ylabel("Values")
plt.title("Sales, Profit, Quantity, and Discount (First 30 Orders)")
plt.legend()
plt.grid(True)
plt.show()
```

### Summary

Adding **axis labels, titles, and legends** is a crucial step in data visualization because they transform raw plots into meaningful insights. Without labels, a plot might look visually appealing but fail to communicate the actual context. Axis labels clarify what data is being represented on each axis — for example, distinguishing between “Sales Amount ($)” and “Number of Orders.” Titles summarize the entire chart in a single line, making it easier for the viewer to instantly grasp the purpose of the visualization. Legends are particularly important when plotting multiple data series on the same graph. They prevent confusion by clearly mapping line colors, styles, or markers to the respective variables. Matplotlib provides both Pyplot and Object-Oriented methods to add these elements, giving flexibility depending on the workflow. Additionally, customization options like font size, color, placement, and gridlines allow for professional presentation-quality plots. In AI/ML or data analysis projects, clear and well-labeled plots are essential when explaining trends, comparing variables, or showcasing results to clients or teammates. Thus, axis labels, titles, and legends serve as the backbone of **storytelling with data**, ensuring that visualizations are not only correct but also interpretable and impactful.
