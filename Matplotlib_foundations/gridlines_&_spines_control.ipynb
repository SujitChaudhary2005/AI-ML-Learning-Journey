# Gridlines and Spines Control in Matplotlib

When we create a plot, the default axes already come with ticks, labels, and sometimes faint gridlines depending on the style. However, in professional data visualization, we often want to **customize gridlines** for better readability and **control spines** (the borders around the plot) to make the plot cleaner and more visually appealing.

### What  is Gridlines in Matplotlib ?

Gridlines are horizontal and vertical reference lines across the plot that make it easier to compare data values. By default, gridlines are usually off in Matplotlib, but we can easily enable and customize them.

- Enable Gridlines:

```python
import matplotlib.pyplot as plt

plt.grid(True)
```

- Customize Gridlines:

We can control line style, color, and width.

```python
plt.grid(True, linestyle='--', color='gray', linewidth=0.7)
```

- Axis-specific Gridlines:

We can choose to show only x-axis or y-axis gridlines.

```python
plt.grid(axis='y')  # Only horizontal gridlines
plt.grid(axis='x')  # Only vertical gridlines
```

### What is Spines in Matplotlib ?

Spines are the borders (lines) around the plot area — top, bottom, left, and right. By default, all spines are visible, but sometimes we want to **remove, shift, or style them** to create cleaner plots.

- Access Spines:

```python
ax = plt.gca()
ax.spines
```

- Remove a Spine:

```python
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
```

- Change Spine Color and Width:

```python
ax.spines['left'].set_color('blue')
ax.spines['left'].set_linewidth(2)
```

- Move Spine Position:

```python
ax.spines['left'].set_position(('outward', 10))  # Move 10 points outward
ax.spines['bottom'].set_position(('data', 0))   # Align with y=0
```

### Example

```python
import pandas as pd

# Sample dataset
df = pd.read_csv("data/Superstore.csv", encoding='latin-1')

# Plot Sales for first 50 rows
plt.plot(df['Sales'][:50], label="Sales", color="blue")

# Add gridlines
plt.grid(True, linestyle='--', color='gray', linewidth=0.7)

# Get current axes
ax = plt.gca()

# Customize spines
ax.spines['top'].set_visible(False)     # Hide top border
ax.spines['right'].set_visible(False)   # Hide right border
ax.spines['left'].set_color("red")      # Left spine in red
ax.spines['bottom'].set_linewidth(2)    # Thicker bottom spine

plt.title("Sales with Custom Gridlines & Spines")
plt.xlabel("Order Index")
plt.ylabel("Sales")
plt.legend()
plt.show()
```

Q1. Plot **Sales for the first 100 rows** with gridlines enabled and styled as dashed gray lines.

```python
plt.figure(figsize=(6,4))
plt.plot(df["Sales"].head(100), color="blue")
plt.title("Sales with Gridlines")
plt.xlabel("Order Index")
plt.ylabel("Sales")
plt.grid(True, linestyle="--", color="gray")
plt.show()
```

Q2. Plot **Profit for the first 50 rows** with only horizontal gridlines and no vertical ones.

```python
plt.figure(figsize=(6,4))
plt.plot(df["Profit"].head(50), color="red")
plt.title("Profit with Horizontal Gridlines Only")
plt.xlabel("Order Index")
plt.ylabel("Profit")
plt.grid(axis="y", linestyle="--", color="black")
plt.show()
```

Q3. Hide the top and right spines in a plot of Quantity (first 50 rows).

```python
fig, ax = plt.subplots(figsize=(6,4))
ax.plot(df["Quantity"].head(50), color="green")
ax.set_title("Quantity with Hidden Top and Right Spines")
ax.set_xlabel("Order Index")
ax.set_ylabel("Quantity")
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
plt.show()
```

Q4. Customize the left spine to be green and thicker, and the bottom spine to be moved outward by 15 points in a plot of Discount.

```python
fig, ax = plt.subplots(figsize=(6,4))
ax.plot(df["Discount"].head(50), color="purple")
ax.set_title("Discount with Customized Spines")
ax.set_xlabel("Order Index")
ax.set_ylabel("Discount")
ax.spines["left"].set_color("green")
ax.spines["left"].set_linewidth(2.5)
ax.spines["bottom"].set_position(("outward", 15))
plt.show()
```

Q5. Create a plot of Sales vs Profit where the axes cross at the origin (move spines to position (‘data’, 0)).

```python
fig, ax = plt.subplots(figsize=(6,4))
ax.plot(df["Sales"].head(50), df["Profit"].head(50), marker="o", linestyle="")
ax.set_title("Sales vs Profit with Axes at Origin")
ax.set_xlabel("Sales")
ax.set_ylabel("Profit")

# Move spines
ax.spines["left"].set_position(("data", 0))
ax.spines["bottom"].set_position(("data", 0))
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
plt.show()
```

### Summary

Gridlines and spines are two important elements in Matplotlib that control the readability and style of plots. **Gridlines** act as guides across the plot to help the viewer align data points with tick values on the axes. They are particularly useful when dealing with dense datasets or when comparing values across different scales. By default, gridlines are usually turned off, but enabling them (`plt.grid(True)`) makes plots much easier to interpret. Moreover, gridlines can be styled with custom colors, line styles (`'--'`, `'-.'`, `':'`), thickness, and can be limited to only one axis (x or y).

**Spines**, on the other hand, are the borders surrounding the plotting area. Matplotlib provides full control over them, allowing us to hide spines, change their color and thickness, or even reposition them. For example, hiding the top and right spines creates a more modern, minimalist style often seen in scientific papers and dashboards. Repositioning spines (e.g., moving the x-axis spine to y=0) is useful for creating mathematical-style plots where axes cross at the origin.

Together, gridlines and spines help balance **readability and aesthetics**. Gridlines provide structure for comparing values, while spines define the plot’s frame. By controlling these elements, We can make plots look more professional, reduce clutter, and emphasize the data itself rather than unnecessary borders.
