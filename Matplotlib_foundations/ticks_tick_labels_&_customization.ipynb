# Ticks, Tick Labels, and Customization

### What is Ticks & Tick labels ?

In any visualization, ticks are the small marks along the x-axis and y-axis that indicate scale and reference points. Tick labels are the numeric or textual values associated with those ticks. Customizing ticks and labels helps improve readability, emphasize important points, and make plots presentation-ready.

By default, Matplotlib automatically decides the tick locations and labels, but we can manually control them using functions like `plt.xticks()`, `plt.yticks()`, `ax.set_xticks()`, `ax.set_xticklabels()`, and `ax.tick_params()`.

In data science and AI/ML, customizing ticks becomes useful when:

- Showing time series data (e.g., years, months, hours).
- Working with categorical data (e.g., product categories, regions).
- Emphasizing specific ranges (e.g., zooming into losses in a Profit curve).
- Making publication-quality plots with neat formatting.

### Basic Tick Customization with Pyplot

The **Pyplot interface** in Matplotlib (`plt`) allows quick and direct tick customization using `plt.xticks()` and `plt.yticks()`. These functions let us define both the positions of ticks and their labels. This method is simple and best suited for small-scale, quick customizations.

```python
import matplotlib.pyplot as plt
import pandas as pd

# Load dataset
df = pd.read_csv("data/Superstore.csv", encoding='latin-1')

# Plot Sales for first 50 rows - default ticks
plt.plot(df['Sales'][:50])
plt.title("Default Tick Behavior")
plt.xlabel("Index")
plt.ylabel("Sales ($)")
plt.show()

# Custom ticks and labels
plt.plot(df['Sales'][:50])
plt.title("Custom Tick Labels")
plt.xlabel("Order Index")
plt.ylabel("Sales ($)")
plt.xticks([0,10,20,30,40,50], ["0","10","20","30","40","50"])  # custom x ticks
plt.yticks([0,100,200,300,400], ["0$","100$","200$","300$","400$"])  # custom y labels
plt.grid(True)
plt.show()
```

### Object-Oriented (OO) Interface

In the **Object-Oriented (OO) approach**, customization is done through the `Axes` object. This method provides more flexibility than Pyplot, as you can precisely set tick positions, style, and labels for multiple subplots. Functions like `set_xticks()`, `set_xticklabels()`, and `set_yticks()` are frequently used here.

```python
fig, ax = plt.subplots(figsize=(6,4))

ax.plot(df['Profit'][:50], color="red")

# Set ticks manually
ax.set_xticks([0,10,20,30,40,50])
ax.set_xticklabels(["0","10","20","30","40","50"], fontsize=10, color="blue")

ax.set_yticks([-200,0,200,400,600])
ax.set_yticklabels(["-200$","0$","200$","400$","600$"], fontsize=10, color="green")

# Axis labels and title
ax.set_xlabel("Order Index")
ax.set_ylabel("Profit ($)")
ax.set_title("Profit with Custom Ticks", fontsize=14)

plt.grid(True)
plt.show()
```

### Advanced Customization with `tick_params`

The `tick_params()` function provides fine-grained control over tick appearance such as direction, length, width, color, and rotation. This method is especially useful when preparing **publication-ready plots** or when working with dense time series or categorical plots where default ticks may overlap or clutter.

```python
fig, ax = plt.subplots(figsize=(6,4))

ax.plot(df['Quantity'][:30], marker="o", linestyle="--")

# Customize ticks
ax.tick_params(axis="x", direction="inout", length=10, width=2, colors="purple", rotation=45)
ax.tick_params(axis="y", direction="out", length=8, width=2, colors="orange")

ax.set_title("Quantity with Customized Ticks", fontsize=14)
ax.set_xlabel("Order Index")
ax.set_ylabel("Quantity")

plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
```

### Exercises

Q1. Plot Sales for the first 100 rows with x-axis ticks every 20 units and custom labels like "Day 0", "Day 20", etc.

```python
plt.figure(figsize=(8,5))
plt.plot(df['Sales'][:100], color='blue')
plt.xticks(ticks=range(0, 101, 20), labels=[f"Day {i}" for i in range(0, 101, 20)])
plt.title("Sales for First 100 Rows with Custom X-Ticks")
plt.xlabel("Order Index")
plt.ylabel("Sales ($)")
plt.grid(True)
plt.show()
```

Q2. Plot Profit for the first 50 rows with y-axis ticks at `[-200, 0, 200, 400, 600]` labeled as *Loss / Break-even / Profit*.

```python
plt.figure(figsize=(8,5))
plt.plot(df['Profit'][:50], color='green')
plt.yticks(ticks=[-200, 0, 200, 400, 600], 
           labels=["Loss", "Break-even", "Profit 200", "Profit 400", "Profit 600"])
plt.title("Profit with Custom Y-Ticks")
plt.xlabel("Order Index")
plt.ylabel("Profit ($)")
plt.grid(True)
plt.show()
```

Q3. Create a bar plot of Quantity for the first 10 rows with custom category labels on the x-axis.

```python
plt.figure(figsize=(8,5))
plt.bar(range(10), df['Quantity'][:10], color='purple')
plt.xticks(ticks=range(10), labels=[f"Order {i}" for i in range(1,11)])
plt.title("Quantity for First 10 Orders with Custom Labels")
plt.xlabel("Orders")
plt.ylabel("Quantity")
plt.show()
```

Q4. Plot Discount vs Order Index and rotate x-axis labels by 45 degrees for readability.

```python
plt.figure(figsize=(8,5))
plt.plot(df['Discount'][:50], color='orange')
plt.xticks(rotation=45)
plt.title("Discount vs Order Index (Rotated Labels)")
plt.xlabel("Order Index")
plt.ylabel("Discount")
plt.grid(True)
plt.show()
```

Q5. Use `tick_params()` to style ticks (color = red, length = 8, width = 2, inside direction).

```python
plt.figure(figsize=(8,5))
plt.plot(df['Sales'][:30], color='red')
plt.tick_params(axis='both', colors='red', length=8, width=2, direction='in')
plt.title("Styled Ticks Example")
plt.xlabel("Order Index")
plt.ylabel("Sales ($)")
plt.grid(True)
plt.show()
```

### Summary

Ticks and tick labels are fundamental elements of any plot as they indicate reference points and scale. In Matplotlib, ticks can be customized using the **Pyplot interface** (`plt.xticks()`, `plt.yticks()`) for simple modifications, or through the **Object-Oriented interface** (`ax.set_xticks()`, `ax.set_xticklabels()`) for more flexibility and control. For advanced customization, `tick_params()` allows fine-tuning of tick appearance, including color, length, width, rotation, and direction. Customizing ticks is particularly important in AI/ML projects where results such as accuracy, loss, iterations, or categorical variables must be clearly communicated. Properly chosen tick intervals and labels enhance readability, reduce clutter, and help make plots presentation-ready for reports, dashboards, or publications.
