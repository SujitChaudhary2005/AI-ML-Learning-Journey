# Image Display and Manipulation with imshow()

### What is `imshow()`?

The function **`imshow()`** in Matplotlib is used to display images and 2D data as a grid of pixels (rasterized image). Each value in the data array is mapped to a **color** using a **colormap (cmap)**. While it is often used to show real images (like `.png`, `.jpg`), in **data science and AI/ML**, it is equally powerful for visualizing **matrices, heatmaps, attention maps, and feature maps** in neural networks.

**Key Parameters:**

- **`cmap`** → Defines the colormap (`viridis`, `gray`, `plasma`, etc.).
- **`aspect`** → Controls aspect ratio (`auto`, `equal`).
- **`interpolation`** → Defines how pixel values are smoothed (`nearest`, `bilinear`).
- **`extent`** → Maps array indices to specific x/y coordinates.
- **`origin`** → Controls where (0,0) is placed (`upper` or `lower`).

**Use in AI/ML:**

- Displaying **grayscale images** (MNIST digits, medical scans).
- Showing **heatmaps** (confusion matrix, correlation matrix).
- Visualizing **CNN filters and activation maps**.
- Inspecting **attention weights** in transformer models.

**Pyplot Example**

Displaying a Heatmap of Profit

```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("data/Superstore.csv", encoding='latin-1')

# Create a pivot table of Profit by Category and Region
profit_pivot = df.pivot_table(values="Profit", index="Category", columns="Region", aggfunc="mean")

plt.figure(figsize=(8,6))
plt.imshow(profit_pivot, cmap="coolwarm", aspect="auto")
plt.colorbar(label="Average Profit ($)")
plt.title("Profit Heatmap (Pyplot)")
plt.xticks(range(len(profit_pivot.columns)), profit_pivot.columns, rotation=45)
plt.yticks(range(len(profit_pivot.index)), profit_pivot.index)
plt.show()
```

**OO Example** 

Displaying a Sales Heatmap

```python
fig, ax = plt.subplots(figsize=(8,6))

sales_pivot = df.pivot_table(values="Sales", index="Category", columns="Region", aggfunc="mean")

im = ax.imshow(sales_pivot, cmap="viridis", aspect="auto")
fig.colorbar(im, ax=ax, label="Average Sales ($)")

ax.set_title("Sales Heatmap (OO Interface)")
ax.set_xticks(range(len(sales_pivot.columns)))
ax.set_xticklabels(sales_pivot.columns, rotation=45)
ax.set_yticks(range(len(sales_pivot.index)))
ax.set_yticklabels(sales_pivot.index)

plt.show()
```

### Exercises

Q1. Display a correlation matrix of numerical columns using `imshow()` with `cmap="RdBu"`.

```python
# Correlation matrix of numerical columns
corr = df.corr(numeric_only=True)

plt.figure(figsize=(8,6))
plt.imshow(corr, cmap="RdBu", aspect="auto")
plt.colorbar(label="Correlation")
plt.xticks(range(len(corr.columns)), corr.columns, rotation=45)
plt.yticks(range(len(corr.columns)), corr.columns)
plt.title("Correlation Matrix Heatmap")
plt.show()
```

Q2. Show a 10×10 random matrix using `imshow()` with `cmap="plasma"` and interpolation set to `"nearest"`.

```python
import numpy as np

random_matrix = np.random.rand(10,10)

plt.figure(figsize=(6,6))
plt.imshow(random_matrix, cmap="plasma", interpolation="nearest")
plt.colorbar(label="Value")
plt.title("10x10 Random Matrix (Plasma)")
plt.show()
```

Q3. Create a heatmap of average Discount by Region and Segment.

```python
discount_pivot = df.pivot_table(values="Discount", index="Region", columns="Segment", aggfunc="mean")

plt.figure(figsize=(8,6))
plt.imshow(discount_pivot, cmap="coolwarm", aspect="auto")
plt.colorbar(label="Average Discount")
plt.xticks(range(len(discount_pivot.columns)), discount_pivot.columns, rotation=45)
plt.yticks(range(len(discount_pivot.index)), discount_pivot.index)
plt.title("Average Discount by Region and Segment")
plt.show()
```

Q4. Visualize Quantity distribution by Sub-Category (pivot table + `imshow()`).

```python
quantity_pivot = df.pivot_table(values="Quantity", index="Sub-Category", columns="Region", aggfunc="mean")

plt.figure(figsize=(10,6))
plt.imshow(quantity_pivot, cmap="viridis", aspect="auto")
plt.colorbar(label="Average Quantity")
plt.xticks(range(len(quantity_pivot.columns)), quantity_pivot.columns, rotation=45)
plt.yticks(range(len(quantity_pivot.index)), quantity_pivot.index)
plt.title("Quantity Distribution by Sub-Category and Region")
plt.show()
```

### Summary

The **`imshow()` function in Matplotlib** is a versatile tool for displaying both real images and 2D data as pixel grids. Each value in the data is represented by a color from a chosen colormap, making it especially useful for creating **heatmaps, correlation matrices, and matrix visualizations**. With key parameters like `cmap` (colormap), `aspect` (aspect ratio), `interpolation` (smoothing), and `origin` (pixel positioning), `imshow()` provides flexibility in how data is visualized. In **data science and AI/ML**, it is frequently applied to show **grayscale datasets like MNIST digits, correlation heatmaps, confusion matrices, CNN activation maps, and attention weights** in transformer models. Using `imshow()` with pivot tables from Pandas, we can quickly generate insightful heatmaps to compare metrics such as **Profit by Category and Region, or Discount by Segment and Region**, while random matrices can help understand visual scaling and color usage. Additionally, `imshow()` can be used to load and display actual image files (e.g., `.png`, `.jpg`) for preprocessing tasks in computer vision. The function supports both **Pyplot** and **Object-Oriented (OO)** approaches, giving flexibility for quick plots or more customized visualizations. Overall, `imshow()` is not limited to images—it is a powerful visualization method for turning any structured 2D data into **intuitive, color-coded insights**.
