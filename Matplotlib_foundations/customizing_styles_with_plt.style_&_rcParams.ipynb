# Customizing Styles in Matplotlib (plt.style & rcParams)

### What is Customizing Styles?

Customizing plot styles in **Matplotlib** means adjusting the **visual appearance** of charts to make them **more readable, attractive, and consistent**. This includes controlling:

- **Colors** – e.g., color of lines, markers, or backgrounds.
- **Fonts** – sizes and families of titles, labels, and tick marks.
- **Gridlines** – color, style, and transparency.
- **Markers and lines** – shapes, sizes, and thickness.
- **Overall theme** – via predefined styles like `'ggplot'` or `'seaborn-darkgrid'`.

Good styling **improves comprehension**, ensures **professional presentation**, and highlights **key insights**. In AI/ML, clear visualization can help spot **trends, clusters, and anomalies** in datasets.

Matplotlib provides **two main methods** for customization:

1. **`plt.style`** – Apply **predefined or custom style templates** quickly.
2. **`rcParams`** – Fine-tune **individual plot attributes** like line width, font size, and grid style.

Together, they make plotting **efficient, consistent, and visually appealing**.

### 1. Using `plt.style`

`plt.style` lets us apply a **theme to all plots** with a single line of code.

Matplotlib comes with many **built-in styles**:

```python
import matplotlib.pyplot as plt
plt.style.available
```

Some popular styles:

| Style Name | Description |
| --- | --- |
| `'classic'` | Default Matplotlib style. |
| `'ggplot'` | Inspired by R’s ggplot2; modern, colorful, grid-heavy. |
| `'seaborn'` | Integrates Seaborn aesthetics; smooth, clean, pastel colors. |
| `'fivethirtyeight'` | Inspired by FiveThirtyEight.com charts; minimalistic and bold. |
| `'bmh'` | Scientific style with clear grids. |

**Example**

Scatter plot using `ggplot` style

```python
import pandas as pd
df = pd.read_csv("data/Superstore.csv", encoding='latin-1')

plt.style.use('ggplot')  # Apply ggplot style

plt.figure(figsize=(10,6))
plt.scatter(df['Sales'], df['Profit'], c=df['Discount'], cmap='viridis', s=50)
plt.title("Sales vs Profit (Style: ggplot)")
plt.xlabel("Sales ($)")
plt.ylabel("Profit ($)")
plt.colorbar(label="Discount")
plt.show()
```

Advantages:

- Quick and easy to apply.
- Creates **consistent visual theme** across multiple plots.

### 2. Using `rcParams`

`rcParams` is a dictionary-like object in Matplotlib that stores **default styling parameters**. We can customize **each aspect** of a plot, such as:

- **Figure size**: `plt.rcParams["figure.figsize"] = (12,6)`
- **Line width**: `plt.rcParams["lines.linewidth"] = 2`
- **Font size**: `plt.rcParams["font.size"] = 14`
- **Grid style**: `plt.rcParams["grid.linestyle"] = "--"`

This gives **precise control** over plots, perfect for **presentations, papers, or dashboards**.

**Example**

Line plot with custom `rcParams`

```python
plt.rcParams["figure.figsize"] = (12,6)
plt.rcParams["lines.linewidth"] = 2
plt.rcParams["axes.titlesize"] = 18
plt.rcParams["axes.labelsize"] = 14
plt.rcParams["grid.linestyle"] = ":"
plt.rcParams["grid.color"] = "gray"
plt.rcParams["grid.alpha"] = 0.5

cumulative_sales = df['Sales'].cumsum().values
plt.plot(cumulative_sales, color="purple", marker="o", markersize=5, label="Cumulative Sales")
plt.title("Cumulative Sales Over Orders (Custom Style)")
plt.xlabel("Order Index")
plt.ylabel("Cumulative Sales ($)")
plt.grid(True)
plt.legend()
plt.show()
```

### 3. Combining `plt.style` and `rcParams`

We can **use a style as a base** and then **tweak specific properties**:

```python
plt.style.use("seaborn-v0_8-darkgrid")
plt.rcParams["lines.linewidth"] = 3
plt.rcParams["axes.titlesize"] = 20

plt.scatter(df["Sales"], df["Profit"], c=df["Discount"], cmap="coolwarm", s=50)
plt.title("Sales vs Profit (Custom + Seaborn Style)")
plt.xlabel("Sales ($)")
plt.ylabel("Profit ($)")
plt.colorbar(label="Discount")
plt.show()
```

Advantages:

- Base style gives **good overall theme**.
- `rcParams` adjustments allow **fine-tuning for our exact needs**.

### 4. Customizing Colors, Markers, and Fonts

- **Colors**: Use named colors, hex codes (`'#FF5733'`), or colormaps (`cmap='viridis'`).
- **Markers**: `'o'` = circle, `'s'` = square, `'^'` = triangle, `'x'` = cross.
- **Fonts**: Change family or size using `plt.rcParams["font.family"]` and `plt.rcParams["font.size"]`.

**Example**

Scatter plot with markers and font changes

```python
plt.rcParams["font.family"] = "monospace"
plt.rcParams["axes.labelsize"] = 15
plt.rcParams["axes.titlesize"] = 18

plt.scatter(df["Sales"], df["Profit"], c=df["Discount"], cmap="plasma", s=70, marker="^", edgecolor="black")
plt.title("Sales vs Profit with Custom Fonts & Markers")
plt.xlabel("Sales ($)")
plt.ylabel("Profit ($)")
plt.colorbar(label="Discount")
plt.show()
```

### Exercises

Q1. Apply `'fivethirtyeight'` style and plot cumulative Profit over Orders.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("data/Superstore.csv", encoding='latin-1')

# Apply style
plt.style.use('fivethirtyeight')

# Cumulative Profit
cumulative_profit = df['Profit'].cumsum()

plt.figure(figsize=(10,6))
plt.plot(cumulative_profit, color='green', linewidth=2)
plt.title("Cumulative Profit Over Orders")
plt.xlabel("Order Index")
plt.ylabel("Cumulative Profit ($)")
plt.show()
```

Q2. Use `rcParams` to increase marker size and line width, then plot Sales vs Profit as a scatter plot colored by Discount.

```python
# Customize rcParams
plt.rcParams["lines.linewidth"] = 2
plt.rcParams["lines.markersize"] = 8
plt.rcParams["axes.titlesize"] = 16
plt.rcParams["axes.labelsize"] = 14

plt.figure(figsize=(10,6))
plt.scatter(df["Sales"], df["Profit"], c=df["Discount"], cmap="viridis", s=100, edgecolor="black")
plt.title("Sales vs Profit (Colored by Discount)")
plt.xlabel("Sales ($)")
plt.ylabel("Profit ($)")
plt.colorbar(label="Discount")
plt.show()
```

Q3. Combine `'bmh'` style with custom font size, line width, and grid style for Quantity vs Profit.

```python
# Apply 'bmh' style
plt.style.use('bmh')

# Customize rcParams
plt.rcParams["axes.titlesize"] = 18
plt.rcParams["axes.labelsize"] = 14
plt.rcParams["lines.linewidth"] = 2
plt.rcParams["grid.linestyle"] = "--"
plt.rcParams["grid.color"] = "gray"

plt.figure(figsize=(10,6))
plt.scatter(df["Quantity"], df["Profit"], c=df["Discount"], cmap="coolwarm", s=80, edgecolor="black")
plt.title("Quantity vs Profit (Custom Style with BMH)")
plt.xlabel("Quantity")
plt.ylabel("Profit ($)")
plt.colorbar(label="Discount")
plt.show()
```

Q4. Create a customized **3D scatter plot** using `plt.style` and `rcParams` to show Sales, Profit, and Discount.

```python
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# Apply style
plt.style.use("seaborn-v0_8-darkgrid")

# Customize rcParams
plt.rcParams["axes.titlesize"] = 18
plt.rcParams["axes.labelsize"] = 14
plt.rcParams["lines.linewidth"] = 2

fig = plt.figure(figsize=(10,7))
ax = fig.add_subplot(111, projection='3d')

# 3D Scatter Plot
sc = ax.scatter(df["Sales"], df["Profit"], df["Discount"], 
                c=df["Discount"], cmap="plasma", s=60, edgecolor="black")

ax.set_title("3D Scatter Plot: Sales vs Profit vs Discount")
ax.set_xlabel("Sales ($)")
ax.set_ylabel("Profit ($)")
ax.set_zlabel("Discount")
fig.colorbar(sc, ax=ax, label="Discount")
plt.show()
```
