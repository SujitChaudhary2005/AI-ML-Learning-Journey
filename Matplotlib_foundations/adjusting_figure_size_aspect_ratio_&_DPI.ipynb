# Adjusting Figure Size, Aspect Ratio, and DPI in Matplotlib

When plotting with Matplotlib, sometimes the default figure size (6.4 × 4.8 inches) doesn’t fit our needs. For example, if we’re plotting many data points, the plot may look crowded. Similarly, for presentations or publications, we may need high-resolution images with specific sizes. This is where customizing **figure size, aspect ratio, and DPI (dots per inch)** becomes important.

- **Figure Size**: The dimensions of the entire figure, set in inches (`width × height`).
- **Aspect Ratio**: The ratio of width to height of the figure. Changing it makes the plot appear stretched or compressed.
- **DPI (Dots Per Inch)**: Controls the resolution of the figure. Higher DPI gives sharper images (useful for printing/publishing).

### Adjusting Figure Size

By default, Matplotlib uses a smaller figure. Here we increase the size to make the plot clearer for presentations.

**Example**

```python
import matplotlib.pyplot as plt
import pandas as pd

# Load dataset
df = pd.read_csv("data/Superstore.csv", encoding='latin-1')

plt.figure(figsize=(10, 6))  # width=10 inches, height=6 inches
plt.plot(df["Sales"][:100], label="Sales", color="blue")
plt.title("Sales (First 100 Rows) - Larger Figure")
plt.xlabel("Order Index")
plt.ylabel("Sales ($)")
plt.legend()
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
```

### Changing Aspect Ratio

A wide aspect ratio helps when we want to highlight trends across a long x-axis (like time series data).

**Example**

```python
plt.figure(figsize=(12, 4))  # Much wider than tall
plt.plot(df["Profit"][:100], color="red", label="Profit")
plt.title("Profit (Wide Aspect Ratio)")
plt.xlabel("Order Index")
plt.ylabel("Profit ($)")
plt.legend()
plt.grid(True, linestyle=":")
plt.show()
```

### High DPI for Sharper Plots

Increasing DPI makes the plot sharper and more detailed, which is important when exporting figures for reports, posters, or research papers.

**Example**

```python
plt.figure(figsize=(8, 5), dpi=150)  # Default DPI is 100
plt.plot(df["Quantity"][:50], marker="o", color="green", label="Quantity")
plt.title("Quantity (High DPI Example)")
plt.xlabel("Order Index")
plt.ylabel("Quantity")
plt.legend()
plt.grid(True, alpha=0.5)
plt.show()
```

### Multiple Figures with Different Sizes

The same data can look very different with varying figure sizes. Choose sizes based on audience (presentation, notebook, or report).

**Example**

```python
plt.figure(figsize=(6, 4))
plt.plot(df["Discount"][:50], color="purple")
plt.title("Discount - Default Size")
plt.show()

plt.figure(figsize=(12, 8))
plt.plot(df["Discount"][:50], color="purple")
plt.title("Discount - Larger Size")
plt.show()
```

### Exporting High-Resolution Plots

When saving plots, we can specify a higher DPI for publication-quality images. The plot will look sharper when zoomed in or printed.

**Example**

```python
fig = plt.figure(figsize=(8, 6), dpi=200)
plt.plot(df["Sales"][:50], color="orange", label="Sales")
plt.title("High Resolution Plot (For Export)")
plt.xlabel("Order Index")
plt.ylabel("Sales ($)")
plt.legend()
plt.grid(True, linestyle="--")
plt.savefig("sales_high_res.png", dpi=200)  # Save as PNG
plt.show()
```

### Summary

Customizing **figure size, aspect ratio, and DPI** in Matplotlib is essential for making plots more readable, professional, and presentation-ready.

- **Figure Size** determines the overall width and height of the plot (in inches). A larger figure prevents crowding when plotting many data points, while smaller figures are useful for quick previews.
- **Aspect Ratio** is the ratio of width to height. A wide aspect ratio is best for long time-series data, while a tall figure emphasizes vertical variations.
- **DPI (Dots Per Inch)** controls the resolution of the figure. Higher DPI produces sharper images, which are crucial for exporting plots for publications, research papers, or presentations.

By combining these settings, we can tailor plots for different contexts—small figures for notebooks, wide figures for trend visualization, tall figures for emphasizing vertical patterns, and high-DPI figures for polished reports. Additionally, saving plots with high resolution ensures they remain clear when zoomed or printed.

In short, adjusting **size, ratio, and DPI** makes visualizations not only correct but also professional, clear, and adaptable to different audiences.
