# Working with Multiple Figures and Axes (Subplots)

### What Are Figures and Axes in Matplotlib?

In Matplotlib, every visualization is built around two main objects:

- **Figure** → The overall window or canvas that holds the entire visualization.
- **Axes** → The actual plotting area inside the figure (like one chart inside the window).

A single **Figure** can contain **multiple Axes (subplots)**, allowing us to visualize **different datasets or comparisons** side by side.

This is especially useful when we need to:

- Compare multiple variables in one view.
- Create dashboards with multiple charts.
- Separate plots logically but keep them in one figure for presentation.

### Creating Multiple Figures

If we want **completely separate plots**, we can create multiple figures:

```python
import matplotlib.pyplot as plt
import pandas as pd

# Load dataset
df = pd.read_csv("data/Superstore.csv", encoding='latin-1')

plt.figure(figsize=(6,4))
plt.plot(df['Sales'][:50], label="Sales")
plt.title("Figure 1 - Sales")
plt.legend()

plt.figure(figsize=(6,4))
plt.plot(df['Profit'][:50], label="Profit", color='red')
plt.title("Figure 2 - Profit")
plt.legend()

plt.show()
```

Here, `plt.figure()` starts a **new figure**.

### Creating Subplots

Instead of multiple figures, we can create **subplots within a single figure**.

Using `plt.subplot()`

```python
plt.subplot(2,1,1)  # (rows, cols, position)
plt.plot(df['Sales'][:50], label="Sales")
plt.title("Sales")

plt.subplot(2,1,2)
plt.plot(df['Profit'][:50], color='red', label="Profit")
plt.title("Profit")

plt.tight_layout()
plt.show()
```

This creates **2 rows, 1 column** of subplots.
Using `plt.subplots()` (Recommended)

```python
fig, axes = plt.subplots(2, 2, figsize=(10,8))

# Top-left
axes[0,0].plot(df['Sales'][:50], color='blue')
axes[0,0].set_title("Sales")

# Top-right
axes[0,1].plot(df['Profit'][:50], color='red')
axes[0,1].set_title("Profit")

# Bottom-left
axes[1,0].bar(range(10), df['Quantity'][:10], color='green')
axes[1,0].set_title("Quantity")

# Bottom-right
axes[1,1].hist(df['Discount'][:50], bins=10, color='purple')
axes[1,1].set_title("Discount Distribution")

plt.tight_layout()
plt.show()
```

This creates a **2x2 grid of plots** inside one figure.

### Why It Matters in AI/ML and Data Science

- Helps compare **multiple features** (e.g., Sales vs Profit vs Quantity).
- Useful for monitoring **training and validation losses** side by side.
- Allows **model performance metrics** (accuracy, confusion matrix, feature importances) to be displayed together.
- Makes dashboards and reports **compact and professional**.

### Exercises

Q1. Create two separate figures: one plotting Sales (first 100 rows) and another plotting Profit (first 100 rows).

```python
# Sales plot
plt.figure(figsize=(6,4))
plt.plot(df['Sales'][:100], color='blue', label='Sales')
plt.title("Sales (First 100 Rows)")
plt.xlabel("Index")
plt.ylabel("Sales")
plt.legend()
plt.grid(True)
plt.show()

# Profit plot
plt.figure(figsize=(6,4))
plt.plot(df['Profit'][:100], color='red', label='Profit')
plt.title("Profit (First 100 Rows)")
plt.xlabel("Index")
plt.ylabel("Profit")
plt.legend()
plt.grid(True)
plt.show()
```

Q2. Use `plt.subplot()` to create a 1x2 layout: Sales in the first plot, Profit in the second.

```python
plt.figure(figsize=(10,4))

# Subplot 1 (Sales)
plt.subplot(1,2,1)
plt.plot(df['Sales'][:100], color='blue')
plt.title("Sales (First 100 Rows)")
plt.xlabel("Index")
plt.ylabel("Sales")
plt.grid(True)

# Subplot 2 (Profit)
plt.subplot(1,2,2)
plt.plot(df['Profit'][:100], color='red')
plt.title("Profit (First 100 Rows)")
plt.xlabel("Index")
plt.ylabel("Profit")
plt.grid(True)

plt.tight_layout()
plt.show()
```

Q3. Use `plt.subplots(2,2)` to create 4 subplots showing Sales, Profit, Quantity, and Discount.

```python
fig, axes = plt.subplots(2,2, figsize=(10,8))

# Sales
axes[0,0].plot(df['Sales'][:100], color='blue')
axes[0,0].set_title("Sales")
axes[0,0].set_xlabel("Index")
axes[0,0].set_ylabel("Sales")
axes[0,0].grid(True)

# Profit
axes[0,1].plot(df['Profit'][:100], color='red')
axes[0,1].set_title("Profit")
axes[0,1].set_xlabel("Index")
axes[0,1].set_ylabel("Profit")
axes[0,1].grid(True)

# Quantity
axes[1,0].plot(df['Quantity'][:100], color='green')
axes[1,0].set_title("Quantity")
axes[1,0].set_xlabel("Index")
axes[1,0].set_ylabel("Quantity")
axes[1,0].grid(True)

# Discount
axes[1,1].plot(df['Discount'][:100], color='orange')
axes[1,1].set_title("Discount")
axes[1,1].set_xlabel("Index")
axes[1,1].set_ylabel("Discount")
axes[1,1].grid(True)

plt.tight_layout()
plt.show()
```

Q4. Try comparing Sales and Profit in the same figure but with two different y-axes.

```python
fig, ax1 = plt.subplots(figsize=(8,5))

# Plot Sales on left y-axis
ax1.plot(df['Sales'][:100], color='blue', label='Sales')
ax1.set_xlabel("Index")
ax1.set_ylabel("Sales", color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

# Plot Profit on right y-axis
ax2 = ax1.twinx()
ax2.plot(df['Profit'][:100], color='red', label='Profit')
ax2.set_ylabel("Profit", color='red')
ax2.tick_params(axis='y', labelcolor='red')

# Title
plt.title("Comparison: Sales vs Profit (Two Y-Axes)")
fig.tight_layout()
plt.show()
```

### Summary

Working with **multiple figures and subplots** in Matplotlib allows for efficient comparison and organization of visualizations. A **Figure** is the overall canvas, while **Axes** are the individual plotting areas.

- `plt.figure()` creates separate figures.
- `plt.subplot()` divides a figure into smaller regions.
- `plt.subplots()` (recommended) gives full control by returning a **Figure** and multiple **Axes** objects.

In practice, subplots make it easier to compare trends (e.g., Sales vs Profit vs Quantity), display multiple metrics together, and create **professional dashboards**. In AI/ML, this is especially useful for visualizing **training progress, model comparisons, and dataset analysis** side by side.
