# Working with Text in Plots

### What is Text in Plots?

In data visualization, **text elements** are critical for making plots understandable and informative. Besides axes labels and titles, we often need to add **custom text, annotations, and labels** to highlight important aspects of our data. Matplotlib provides powerful functions such as:

- `plt.text()` (Pyplot) or `ax.text()` (OO) → add custom text at specific (x, y) positions.
- `plt.annotate()` (Pyplot) or `ax.annotate()` (OO) → add annotations with arrows pointing to important data points.
- Text properties: font size, color, rotation, alignment, and style (bold, italic).

In **AI/ML workflows**, text in plots is often used to:

- Highlight **best/worst performing models** in comparison charts.
- Annotate **training progress** (e.g., accuracy milestones).
- Label **clusters in scatter plots** (e.g., customer segmentation).
- Add **context to error analysis plots**, such as pointing out anomalies.

Text improves readability, storytelling, and makes visualizations presentation-ready.

**Pyplot Example: Adding Text in a Sales Trend Plot**

```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("data/Superstore.csv", encoding='latin-1')

# Take first 50 rows for clarity
df_subset = df.head(50)

plt.figure(figsize=(12,6))
plt.plot(df_subset.index, df_subset["Sales"], marker="o", color="blue", label="Sales")

# Add custom text
plt.text(10, 400, "Sales growing here", fontsize=12, color="green", rotation=15)
plt.text(30, 100, "Dip in Sales", fontsize=12, color="red", rotation=-10)

plt.title("Sales Trend with Custom Text (Pyplot)")
plt.xlabel("Order Index")
plt.ylabel("Sales ($)")
plt.legend()
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
```

**OO Example**

```python
fig, ax = plt.subplots(figsize=(10,5))

df_subset = df.head(50)
ax.plot(df_subset.index, df_subset["Profit"], marker="s", color="purple", label="Profit")

# Add text labels
ax.text(5, 200, "High Profit Zone", fontsize=12, color="darkgreen")
ax.text(25, -200, "Loss Detected", fontsize=12, color="red")

ax.set_title("Profit with Text Annotations (OO Interface)")
ax.set_xlabel("Order Index")
ax.set_ylabel("Profit ($)")
ax.legend()
ax.grid(True, linestyle="--", alpha=0.6)

plt.show()
```

### Exercises

Q1. Add a text label to show the maximum Sales in the first 100 rows.

```python
df_subset = df.head(100)
max_idx = df_subset["Sales"].idxmax()
max_val = df_subset["Sales"].max()

plt.figure(figsize=(12,6))
plt.plot(df_subset.index, df_subset["Sales"], color="blue", marker="o")

plt.text(max_idx, max_val+50, f"Max Sales: {max_val:.2f}", fontsize=12, color="red")
plt.title("Sales with Maximum Labeled")
plt.xlabel("Order Index")
plt.ylabel("Sales ($)")
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
```

Q2. Annotate the lowest Profit in the first 50 rows.

```python
df_subset = df.head(50)
min_idx = df_subset["Profit"].idxmin()
min_val = df_subset["Profit"].min()

plt.figure(figsize=(12,6))
plt.plot(df_subset.index, df_subset["Profit"], marker="s", color="purple")

plt.annotate("Lowest Profit", xy=(min_idx, min_val),
             xytext=(min_idx+5, min_val-100),
             arrowprops=dict(facecolor="black", arrowstyle="->"),
             fontsize=12, color="red")

plt.title("Profit Trend with Lowest Value Annotated")
plt.xlabel("Order Index")
plt.ylabel("Profit ($)")
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
```

Q3. Add rotated text labels for each Category in a bar chart of average Sales.

```python
category_sales = df.groupby("Category")["Sales"].mean()

plt.figure(figsize=(8,5))
plt.bar(category_sales.index, category_sales, color="lightblue")

for i, val in enumerate(category_sales):
    plt.text(i, val+20, f"{val:.0f}", ha="center", rotation=45, fontsize=10)

plt.title("Average Sales per Category with Text Labels")
plt.xlabel("Category")
plt.ylabel("Sales ($)")
plt.show()
```

Q4. Use bold and italic text in a Discount vs Sales scatter plot.

```python
plt.figure(figsize=(8,5))
plt.scatter(df["Discount"].head(100), df["Sales"].head(100), color="orange")

plt.text(0.2, 500, "Moderate Discounts\n*Boost Sales*", fontsize=12, fontstyle="italic", color="green")
plt.text(0.7, 200, "High Discounts\n**Low Profit**", fontsize=12, fontweight="bold", color="red")

plt.title("Discount vs Sales with Styled Text")
plt.xlabel("Discount")
plt.ylabel("Sales ($)")
plt.show()
```

Q5. Add multiline text inside a plot to explain a trend.

```python
df_subset = df.head(60)

plt.figure(figsize=(12,6))
plt.plot(df_subset.index, df_subset["Sales"], marker="o", color="blue")

plt.text(20, 300, "Notice the steady increase\nin sales over this period.\nThis may indicate a seasonal trend.",
         fontsize=11, color="darkblue", bbox=dict(facecolor="lightyellow", alpha=0.5))

plt.title("Sales with Multiline Text Explanation")
plt.xlabel("Order Index")
plt.ylabel("Sales ($)")
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
```

### Summary

**Working with text in plots** is not just about aesthetics — it’s about communication. Labels, titles, and custom text guide the viewer’s attention and make data insights clear. In the **Superstore dataset**, text helps highlight spikes in sales, dips in profit, or anomalies in discounts.

In **AI/ML**, text annotations are indispensable when:

- Highlighting **accuracy milestones** during training.
- Annotating **loss spikes** to detect instability.
- Adding **cluster names** in PCA/t-SNE visualizations.
- Making plots **presentation-ready** by labeling important findings.

Good text usage transforms raw numbers into **clear insights**. Without text, a chart may look pretty but fail to communicate. With effective text placement, plots become powerful storytelling tools.
