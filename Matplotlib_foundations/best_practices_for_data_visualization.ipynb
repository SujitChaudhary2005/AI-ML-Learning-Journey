# Best Practices for Data Visualization

### What is Data Visualization Best Practices?

Data visualization is the process of **representing data visually** to make it easier to **analyze, interpret, and communicate insights**. While creating charts and graphs is straightforward in tools like Matplotlib, Pandas, or Seaborn, **following best practices ensures that visualizations are accurate, clear, and effective**. Poorly designed plots can **mislead viewers**, obscure trends, or fail to communicate the intended message.

Best practices cover several aspects of visualization:

- **Clarity:** The chart should clearly convey the message without confusion.
- **Accuracy:** Avoid distorted scales or misleading representations.
- **Aesthetics:** Use consistent colors, fonts, and styles to make plots visually appealing.
- **Simplicity:** Avoid unnecessary decorations or clutter that distract from the data.
- **Consistency:** Keep scales, labels, and colors consistent across multiple plots.
- **Accessibility:** Ensure readability for color-blind audiences and maintain proper font sizes.

In **AI/ML and data science**, adhering to best practices is critical because visualizations are often used to **analyze model performance, compare datasets, or communicate insights to stakeholders**.

**Examples**

1. **Choose the Right Plot Type**

Use a plot that matches the type of data:

- **Line plots:** Trends over time (e.g., sales over months).
- **Bar charts:** Compare categories (e.g., profit by region).
- **Histograms:** Understand distributions (e.g., customer ages).
- **Scatter plots:** Show relationships (e.g., sales vs profit).

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(5)
y = [100, 200, 150, 300, 250]

plt.bar(x, y, color="skyblue")
plt.title("Sales by Region")
plt.xlabel("Region")
plt.ylabel("Sales ($)")
plt.show()
```

2. **Use Labels, Titles, and Legends**

Always label axes, provide a title, and add a legend when needed.

```python
plt.plot(x, y, marker="o", label="Sales")
plt.title("Sales Trend")
plt.xlabel("Region Index")
plt.ylabel("Sales ($)")
plt.legend()
plt.show()
```

3. **Maintain Proper Scale**

Avoid distorting data with misleading scales. Start axes at zero for bar charts when comparing magnitudes.

```python
# Bad example: y-axis not starting at 0
plt.bar(x, y, color="orange")
plt.ylim(100, 300)  # misleading scale
plt.show()

# Good example:
plt.bar(x, y, color="orange")
plt.ylim(0, 350)  # accurate representation
plt.show()
```

4. **Use Color Wisely**

- Use distinct colors for clarity.
- Avoid excessive colors.
- Consider color-blind friendly palettes.

```python
categories = ["A", "B", "C"]
values = [10, 20, 15]

plt.bar(categories, values, color=["#1f77b4", "#ff7f0e", "#2ca02c"])
plt.show()
```

5. **Simplify and Remove Clutter**

Avoid gridlines or extra decorations that do not add information.

```python
plt.plot(x, y, color="green")
plt.title("Clean Line Plot")
plt.grid(False)  # remove clutter
plt.show()
```

6. **Use Subplots for Comparison**

Compare multiple related plots side by side.

```python
fig, axs = plt.subplots(1, 2, figsize=(10,4))
axs[0].bar(x, y, color="blue")
axs[0].set_title("Bar Chart")
axs[1].plot(x, y, color="red")
axs[1].set_title("Line Chart")
plt.tight_layout()
plt.show()
```

7. **Highlight Key Data**

Emphasize important points or trends.

```python
plt.plot(x, y, color="blue")
plt.scatter([3], [300], color="red", s=100, label="Peak Sales")
plt.title("Sales Trend Highlight")
plt.xlabel("Region")
plt.ylabel("Sales")
plt.legend()
plt.show()
```

### Exercises

Q1. Create a bar chart showing total Sales by Category, with proper labels and title.

```python
import pandas as pd
df = pd.read_csv("data/Superstore.csv", encoding='latin-1')

sales_by_cat = df.groupby("Category")["Sales"].sum()

plt.bar(sales_by_cat.index, sales_by_cat.values, color="skyblue")
plt.title("Total Sales by Category")
plt.xlabel("Category")
plt.ylabel("Sales ($)")
plt.show()
```

Q2. Plot Sales vs Profit as a scatter plot, highlight points where Profit > 200.

```python
plt.scatter(df["Sales"], df["Profit"], alpha=0.6)
highlight = df["Profit"] > 200
plt.scatter(df["Sales"][highlight], df["Profit"][highlight], color="red", label="Profit > 200")
plt.title("Sales vs Profit")
plt.xlabel("Sales ($)")
plt.ylabel("Profit ($)")
plt.legend()
plt.show()
```

Q3. Generate a line chart for monthly Sales, using a clean style without unnecessary gridlines.

```python
monthly_sales = df.groupby("Order Date")["Sales"].sum().head(12)
plt.plot(monthly_sales, color="green", linewidth=2)
plt.title("Monthly Sales (Clean Style)")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.grid(False)
plt.show()
```

Q4. Use subplots to compare Sales and Profit distributions in two histograms side by side.

```python
fig, axs = plt.subplots(1, 2, figsize=(10,4))
axs[0].hist(df["Sales"], bins=20, color="blue", alpha=0.7)
axs[0].set_title("Sales Distribution")
axs[1].hist(df["Profit"], bins=20, color="orange", alpha=0.7)
axs[1].set_title("Profit Distribution")
plt.tight_layout()
plt.show()
```

Q5. Apply color-blind friendly colors in a bar chart showing Quantity by Region.

```python
regions = df["Region"].unique()
quantity_by_region = df.groupby("Region")["Quantity"].sum()
plt.bar(regions, quantity_by_region.values, color=["#0072B2", "#D55E00", "#F0E442", "#009E73"])
plt.title("Quantity by Region (Color-blind Friendly)")
plt.show()
```

Key principles include **choosing the correct plot type, using labels and titles, maintaining proper scales, simplifying visuals, applying consistent styles, and using appropriate colors**. In AI/ML and data analytics, adhering to these practices is critical because visualizations often inform decisions, demonstrate model performance, or highlight trends in datasets. Clean, clear, and well-labeled visuals not only make our analysis more trustworthy but also enhance **readability, accessibility, and interpretability**. Using subplots, highlighting key data, and applying color-blind friendly palettes further improves communication.
