# Creating Our First Plot: plt.plot() Basics

### What is `plt.plot()`?

The `plt.plot()` function is the most fundamental plotting function in Matplotlib’s **Pyplot API**. It is used to create line plots, which connect data points with straight lines. Line plots are ideal for showing **trends, patterns, and continuous relationships** between variables.

At its core, `plt.plot()` takes two arguments:

```python
plt.plot(x_values, y_values)
```

- `x_values` → values on the x-axis (independent variable)
- `y_values` → values on the y-axis (dependent variable)

If only `y_values` are provided, Matplotlib automatically assumes `x = [0, 1, 2, ...]`.

**Example**

````python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.plot(x, y)
plt.show()
````

This will create a simple straight line passing through the given points.

### Why is it important?

Line plots are one of the **most common types of visualizations** in data analysis and AI/ML because:

- They show **trends over time** (e.g., sales growth, stock prices, accuracy during training).
- They highlight **relationships between variables**.
- They are useful for comparing multiple series on the same graph.

In business datasets like **Superstore**, line plots can visualize **sales trends**, **profit trends**, or changes in discounts over orders. In ML, they are used to monitor **loss vs. epochs** or **accuracy vs. epochs** during model training.

### The Basics of `plt.plot()`

The function has a rich set of parameters to control appearance:

````python
plt.plot(x, y, color='blue', linestyle='--', marker='o', label='Line A')
````

- `color` → line color (`'r'`, `'g'`, `'b'`, `'black'`, or hex codes like `'#FF5733'`).
- `linestyle` → line style (`'-'` solid, `'--'` dashed, `':'` dotted, `'-.'` dash-dot).
- `marker` → symbol at each data point (`'o'` circle, `'s'` square, `'^'` triangle, etc.).
- `label` → label for the legend.

### Example with Superstore Dataset

````python
import pandas as pd
import matplotlib.pyplot as plt

# Load Superstore dataset
df = pd.read_csv("data/Superstore.csv", encoding='latin-1')

# Take first 50 rows of Sales
sales_data = df['Sales'][:50]

# Create a simple line plot
plt.figure(figsize=(10,5))
plt.plot(sales_data, color='blue', marker='o', linestyle='-', label='Sales')
plt.title("Sales for First 50 Orders")
plt.xlabel("Order Index")
plt.ylabel("Sales")
plt.legend()
plt.grid(True)
plt.show()
````

This plot shows how **sales vary across the first 50 rows/orders** in the dataset.

### Exercises

Q1. Load the Superstore dataset and plot Sales for the first 100 rows using `plt.plot()` .

````python
import pandas as pd

df = pd.read_csv("data/Superstore.csv", encoding="latin1")

plt.figure(figsize=(8,5))
plt.plot(df['Sales'][:100])
plt.title("Sales for First 100 Rows")
plt.xlabel("Index")
plt.ylabel("Sales")
plt.show()
````

Q2. Plot Profit for the first 100 rows with `marker='x'` and color `'green'`.

````python
plt.figure(figsize=(8,5))
plt.plot(df['Profit'][:100], marker='x', color='green')
plt.title("Profit for First 100 Rows")
plt.xlabel("Index")
plt.ylabel("Profit")
plt.show()
````

Q3. On the same graph, plot both Sales and Profit for the first 50 rows.

````python
plt.figure(figsize=(8,5))
plt.plot(df['Sales'][:50], label="Sales", color='blue')
plt.plot(df['Profit'][:50], label="Profit", color='orange')
plt.title("Sales vs Profit (First 50 Rows)")
plt.xlabel("Index")
plt.ylabel("Value")
plt.legend()
plt.show()
````

Q4. Try changing line styles (`--`, `:`, `-.`) and markers (`o`, `s`, `^`) to see the differences.

````python
plt.figure(figsize=(8,5))
plt.plot(df['Sales'][:30], linestyle='--', marker='o', label="Sales --o")
plt.plot(df['Profit'][:30], linestyle=':', marker='s', label="Profit :s")
plt.plot(df['Quantity'][:30], linestyle='-.', marker='^', label="Quantity -.^")
plt.title("Line Styles and Markers Demo")
plt.xlabel("Index")
plt.ylabel("Values")
plt.legend()
plt.show()
````

Q5. Add a title, axis labels, grid, and legend to make the plot presentation-ready.

````python
plt.figure(figsize=(10,6))
plt.plot(df['Sales'][:50], label="Sales", linestyle='-', marker='o', color='blue')
plt.plot(df['Profit'][:50], label="Profit", linestyle='--', marker='x', color='red')
plt.title("Sales vs Profit (Presentation Ready)", fontsize=14, fontweight='bold')
plt.xlabel("Order Index", fontsize=12)
plt.ylabel("Value ($)", fontsize=12)
plt.legend(loc="upper left")
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
````

### Summary

The `plt.plot()` function is the **core building block of plotting in Matplotlib**. It allows quick creation of line plots, which are essential for showing trends, relationships, and comparisons between variables. With parameters like `color`, `linestyle`, `marker`, and `label`, we can fully customize how the lines look. In business datasets like Superstore, `plt.plot()` helps visualize how sales, profit, or discount values change over time or across transactions. In AI/ML, it is indispensable for monitoring training progress through accuracy and loss curves. Mastering `plt.plot()` sets the foundation for more advanced visualizations.
