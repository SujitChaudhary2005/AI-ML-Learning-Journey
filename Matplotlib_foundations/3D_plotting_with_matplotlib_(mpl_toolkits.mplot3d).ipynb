# 3D Plotting with Matplotlib (mpl_toolkits.mplot3d)

### What is 3D Plotting?

3D plotting is the process of creating **three-dimensional visualizations** where data is represented using three axes: **X, Y, and Z**. While traditional 2D plots (like line plots or scatter plots) can only show relationships between two variables, 3D plots allow us to explore **interactions and dependencies among three variables at once**.

In practice, this means that instead of just drawing a flat chart, we can represent data in a **spatial format** where each point, line, or surface exists in 3D space. This is extremely valuable for datasets where relationships are too complex to be fully understood in 2D.

Matplotlib supports 3D plotting through the **`mpl_toolkits.mplot3d`** module. To enable 3D axes, we create a special axis object with:

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")
```

Once the 3D axis is initialized, we can draw:

- **3D lines** to show trajectories or paths.
- **3D scatter plots** to represent distributions or clusters.
- **3D surfaces and wireframes** to show mathematical functions or model landscapes.

In **AI/ML**, 3D plotting is especially useful because many problems involve **multidimensional relationships**. For example:

- Visualizing **loss functions** over weights to see how optimization works.
- Displaying **clusters** in 3D feature spaces.
- Showing **decision boundaries** for classifiers.
- Plotting **time series with an additional dimension**, such as accuracy vs. epoch vs. learning rate.

Thus, 3D plotting is not only about aesthetics but also about **revealing insights that would remain hidden in 2D**.

### Common 3D Plot Types

1. 3D Line Plot

A line in 3D space can show the relationship between three variables.

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")

z = np.linspace(0, 10, 100)
x = np.sin(z)
y = np.cos(z)

ax.plot(x, y, z, label="3D Line")
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
ax.set_zlabel("Z-axis")
ax.set_title("3D Line Plot")
plt.show()
```

2. 3D Scatter Plot

Shows individual points in 3D, useful for clustering and distribution visualization.

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")

x = np.random.rand(50)
y = np.random.rand(50)
z = np.random.rand(50)

ax.scatter(x, y, z, c=z, cmap="viridis", s=50)
ax.set_title("3D Scatter Plot")
plt.show()
```

3. 3D Surface Plot

Represents a continuous surface over 2D space, widely used in **AI/ML for loss surfaces**.

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")

X = np.linspace(-5, 5, 50)
Y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(X, Y)
Z = np.sin(np.sqrt(X**2 + Y**2))

ax.plot_surface(X, Y, Z, cmap="plasma", edgecolor="none")
ax.set_title("3D Surface Plot")
plt.show()
```

4. 3D Wireframe Plot

A wire mesh representation of a surface  lighter and faster than surface plot.

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")

X = np.linspace(-5, 5, 30)
Y = np.linspace(-5, 5, 30)
X, Y = np.meshgrid(X, Y)
Z = np.cos(np.sqrt(X**2 + Y**2))

ax.plot_wireframe(X, Y, Z, color="black")
ax.set_title("3D Wireframe Plot")
plt.show()
```

### Exercises

Q1. Create a 3D scatter plot of Sales, Profit, and Discount.

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import pandas as pd

df = pd.read_csv("data/Superstore.csv", encoding='latin-1')

fig = plt.figure(figsize=(10,7))
ax = fig.add_subplot(111, projection="3d")

ax.scatter(df["Sales"], df["Profit"], df["Discount"],
           c=df["Profit"], cmap="viridis", s=40)

ax.set_title("3D Scatter Plot of Sales, Profit, and Discount")
ax.set_xlabel("Sales ($)")
ax.set_ylabel("Profit ($)")
ax.set_zlabel("Discount")
plt.show()
```

Q2. Plot a 3D line of cumulative Sales over Orders.

```python
import numpy as np

df_subset = df.head(200)
cumulative_sales = df_subset["Sales"].cumsum().values
x = np.arange(len(cumulative_sales))
y = cumulative_sales
z = np.linspace(0, 50, len(cumulative_sales))  # artificial depth

fig = plt.figure(figsize=(10,7))
ax = fig.add_subplot(111, projection="3d")

ax.plot(x, y, z, color="blue", linewidth=2)

ax.set_title("3D Line Plot of Cumulative Sales")
ax.set_xlabel("Order Index")
ax.set_ylabel("Cumulative Sales ($)")
ax.set_zlabel("Depth (Epochs)")
plt.show()
```

Q3. Plot a 3D surface of random data representing Sales Trends.

```python
X = np.linspace(-5, 5, 50)
Y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(X, Y)
Z = np.sin(np.sqrt(X**2 + Y**2))  # artificial trend

fig = plt.figure(figsize=(10,7))
ax = fig.add_subplot(111, projection="3d")

surf = ax.plot_surface(X, Y, Z, cmap="plasma")

ax.set_title("3D Surface Plot - Sales Trend Example")
ax.set_xlabel("X Axis")
ax.set_ylabel("Y Axis")
ax.set_zlabel("Trend Value")
fig.colorbar(surf, ax=ax, shrink=0.5, aspect=5)
plt.show()
```

Q4. Visualize Sales vs Profit vs Quantity using a 3D scatter plot with color mapping by Discount.

### Summary

3D plotting with **Matplotlibs `mpl_toolkits.mplot3d`** expands the power of visualization by allowing us to work with data in three dimensions. Unlike 2D plots where we only see relationships across **X and Y**, 3D plots add a **Z-axis**, making it possible to visualize complex relationships, surfaces, and multidimensional structures.

There are different types of 3D plots, each serving specific use cases:

- **3D Line Plots** are great for showing a path or relationship over three axes, for example, movement in space or a time-based trajectory.
- **3D Scatter Plots** let us examine point distributions in 3D space, which is essential in **clustering and classification problems** (e.g., plotting decision boundaries or K-means clusters).
- **Surface Plots** show continuous surfaces where Z depends on X and Y. In AI/ML, this is often used to visualize **loss functions** or optimization landscapes.
- **Wireframe Plots** are a lightweight way of plotting 3D surfaces without filling them, useful for quick inspections.

In real-world data science and AI, 3D plotting is often applied to:

- **Visualizing high-dimensional datasets** by projecting them into 3D.
- **Inspecting model performance** like error surfaces or accuracy landscapes.
- **Analyzing spatial or geographical data** (e.g., terrain plots).
- **Communicating clustering results** by coloring points by class.

Ultimately, 3D visualization is not just about making graphs look cool. It helps researchers and engineers interpret data that cannot be fully understood in 2D space, making it a vital skill in advanced analytics and machine learning.
