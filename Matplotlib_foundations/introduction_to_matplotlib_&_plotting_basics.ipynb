# Introduction to Matplotlib & Plotting Basics

### What is Matplotlib?

Matplotlib is the foundational Python library for **data visualization**. It allows the creation of **high-quality plots, charts, and figures**, making it an essential tool for data analysis, exploration, and communication. From simple line graphs to complex 3D visualizations and interactive dashboards, Matplotlib provides versatile plotting capabilities. It integrates seamlessly with **NumPy and Pandas**, which allows us to visualize arrays and DataFrame data directly. For example, using the **Sample - Superstore dataset**, we can plot sales trends, customer distributions, or profit analysis in just a few lines of code.

Think of Matplotlib as the **“drawing canvas” of Python**:

- We create a **figure** (the entire drawing area).
- Inside it, we add one or more **axes** (subplots where plots live).
- On each axis, we plot our **data**.

### Why is it important in AI/ML & Data Science?

- Data exploration is impossible without visualization — numbers alone don’t reveal patterns.
- We can see **trends, correlations, and outliers** before applying ML models.
- Example: In Titanic dataset, survival rates by age/gender/class become obvious with plots.
- When reporting ML results, plots like **loss curves, accuracy vs. epochs, confusion matrices, ROC curves** are all created using Matplotlib.

### **Matplotlib Architecture**

Matplotlib follows a **layered architecture** consisting of Figures, Axes, and Artists:

1. **Figure** – This is the top-level container, essentially the blank canvas for our plots. A single figure can hold multiple Axes.
2. **Axes** – Represents a single plot within the figure, containing the data points, labels, ticks, and legends. We can have multiple Axes in one Figure, useful for subplots or grids.
3. **Axis** – Each Axes contains an x-axis and y-axis, responsible for scaling and tick labeling.
4. **Artists** – Everything that is drawn on the plot (lines, markers, text, shapes) is an Artist. This allows precise customization at the smallest level.

### **Interfaces in Matplotlib**

- **Pyplot API**: State-machine-like interface that is simple and convenient, similar to MATLAB. Quick for creating basic plots:

```python
import matplotlib.pyplot as plt

plt.plot([1, 2, 3], [4, 5, 6])
plt.title("Simple Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

- **Object-Oriented (OO) API**: Recommended for complex or multiple plots. Offers more control over Figures and Axes:

```python
fig, ax = plt.subplots()

ax.plot([1, 2, 3], [4, 5, 6], label="Line")
ax.set_title("OO Line Plot")
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
ax.legend()
plt.show()
```

### Direct Plotting from Pandas DataFrame

Pandas integrates Matplotlib directly, allowing quick plotting without explicit loops.

```python
import pandas as pd

df = pd.read_csv("data/Superstore.csv", encoding='latin-1')

# Direct plot of Sales column
df['Sales'][:50].plot(kind='line', color='purple', marker='o', title='Sales - First 50 Rows')
plt.xlabel('Index')
plt.ylabel('Sales')
plt.grid(True)
plt.show()
```

This is very convenient for **EDA** and plotting multiple columns quickly.

### Why Learn Matplotlib?

Understanding Matplotlib is crucial for **data exploration, presentation, and AI/ML feature visualization**. Visualizations help spot trends, outliers, and correlations that are critical for **EDA** (Exploratory Data Analysis) and for building predictive models. For example, plotting `Sales` vs `Profit` from the Superstore dataset can reveal which products are most profitable, guiding AI-driven business decisions.

### Exercises

Q1. Load the Superstore dataset and print the first 5 rows.

```python
df = pd.read_csv("data/Superstore.csv", encoding='latin-1')

# Print first 5 rows
print("First 5 rows of the dataset:")
print(df.head())
```

Q2. Pyplot line plot of first 100 Sales values.

```python
# First 100 Sales values
sales_data = df['Sales'][:100]

# Pyplot line plot
plt.figure(figsize=(10,5))
plt.plot(sales_data, color='blue', marker='o', linestyle='-', label='Sales')
plt.title("Sales for First 100 Rows")
plt.xlabel("Index")
plt.ylabel("Sales")
plt.grid(True)
plt.legend()
plt.show()
```

Q3. OO API scatter plot of Discount vs Profit.

```python
fig, ax = plt.subplots(figsize=(8,5))
ax.scatter(df['Discount'], df['Profit'], color='green', label='Profit vs Discount')
ax.set_title("Discount vs Profit")
ax.set_xlabel("Discount")
ax.set_ylabel("Profit")
ax.legend()
ax.grid(True)
plt.show()
```

Q5. Category-wise total Sales bar chart

```python
category_sales = df.groupby('Category')['Sales'].sum()

plt.figure(figsize=(8,5))
plt.bar(category_sales.index, category_sales.values, color=['blue','green','red'])
plt.title("Total Sales by Category")
plt.xlabel("Category")
plt.ylabel("Total Sales")
plt.grid(axis='y')
plt.show()
```

### Summary

Matplotlib is the foundational Python library for data visualization, providing the tools to create high-quality plots, charts, and figures that are essential for data analysis, exploration, and communication. It integrates seamlessly with NumPy and Pandas, allowing direct visualization of arrays and DataFrame columns, which is particularly useful during exploratory data analysis (EDA) and feature inspection for AI/ML projects. Its layered architecture—comprising Figures, Axes, Axis, and Artists—offers a structured approach to plotting, giving both flexibility and precision. The Figure acts as the overall canvas, Axes represent individual plots within the figure, Axis manages the scaling and labeling of x and y axes, and Artists include every visual element, from lines and markers to text and shapes. Matplotlib offers two main interfaces: the Pyplot API, which is a simple, state-machine-like interface suitable for quick plotting, and the Object-Oriented API, which is recommended for complex plots, multiple subplots, and finer control over visualization elements. Mastery of plot styling—such as colors, line styles, markers, legends, gridlines, and labels—is crucial for making plots readable, professional, and presentation-ready. Additionally, Pandas supports direct plotting from DataFrame columns, streamlining the visualization workflow and enabling fast insights without explicit loops. Using datasets like Superstore, one can plot sales trends, profits, discounts, or category-level metrics to discover patterns, spot outliers, and guide AI-driven decisions. Understanding these fundamentals, along with styling and direct plotting, lays the foundation for creating informative, attractive, and actionable visualizations, which are indispensable skills for any data scientist or AI/ML practitioner.
