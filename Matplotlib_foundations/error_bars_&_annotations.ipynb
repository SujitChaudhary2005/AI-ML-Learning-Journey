# Error Bars and Annotations in Matplotlib

### What are Error Bars?

An **Error Bar** is a graphical representation of the variability or uncertainty in data. It gives an indication of how precise a measurement is or how much the data might vary. Typically, error bars extend above and below a data point, showing the possible range of values.

- In **business analytics**, error bars can represent uncertainty in sales forecasts, profit variability across regions, or the margin of error in customer satisfaction surveys.
- In **AI/ML**, error bars are useful when visualizing **model evaluation results** — for example, showing standard deviation of accuracy across multiple runs, or variability in loss values during cross-validation.

Error bars make visualizations more **trustworthy** by clearly communicating **confidence intervals, deviations, or measurement errors**, which raw numbers alone cannot show.

**Example**

```python
import matplotlib.pyplot as plt
import pandas as pd

# Load Superstore dataset
df = pd.read_csv("data/Superstore.csv", encoding='latin-1')

# Group by category
category_sales = df.groupby("Category")["Sales"].mean()
category_std = df.groupby("Category")["Sales"].std()

# Plot with error bars
plt.figure(figsize=(8,5))
plt.bar(category_sales.index, category_sales, 
        yerr=category_std, capsize=5, color='skyblue')

plt.title("Average Sales per Category with Error Bars")
plt.xlabel("Category")
plt.ylabel("Average Sales ($)")
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
```

Here, the **bar chart** shows average sales for each category. The **error bars** (using standard deviation) give an idea of how spread out the sales are within each category. For example, if “Technology” has a larger error bar, it means the sales values fluctuate more compared to “Furniture” or “Office Supplies.” Error bars are especially useful for executives or analysts to understand **data variability** before making decisions.

### What are Annotations?

An **Annotation** in Matplotlib is a way to add explanatory text or symbols directly onto a plot. Unlike titles or axis labels, annotations highlight **specific data points or regions**.

- In **business analysis**, annotations can mark **important sales spikes**, **major losses**, or **seasonal peaks** in customer orders.
- In **AI/ML**, annotations are often used to mark **maximum accuracy achieved**, **epoch of convergence**, or **anomaly detection points**.

Annotations improve storytelling in data visualization. A plot without annotations may show trends, but a plot with targeted labels tells the audience exactly **why** a data point matters.

**Example**

```python
# Take first 100 rows for example
df_subset = df.head(100)

plt.figure(figsize=(10,5))
plt.plot(df_subset.index, df_subset["Profit"], marker='o', label="Profit")

# Add annotation at max profit
max_profit_idx = df_subset["Profit"].idxmax()
max_profit_val = df_subset["Profit"].max()
plt.annotate("Highest Profit", 
             xy=(max_profit_idx, max_profit_val), 
             xytext=(max_profit_idx+5, max_profit_val+200), 
             arrowprops=dict(facecolor="red", arrowstyle="->"))

plt.title("Profit Trend with Annotation")
plt.xlabel("Order Index")
plt.ylabel("Profit ($)")
plt.legend()
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
```

In this plot, the **line graph** shows profit over the first 100 orders. An annotation highlights the **maximum profit point**, making it easy for viewers to quickly notice where the peak occurs. This is helpful in business presentations, where pointing out the highest or lowest values makes storytelling clearer.

### Exercises

Q1. Plot Average Profit per Category with Error Bars (Standard Deviation).

```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("data/Superstore.csv", encoding='latin-1')

category_profit = df.groupby("Category")["Profit"].mean()
category_std = df.groupby("Category")["Profit"].std()

plt.figure(figsize=(8,5))
plt.bar(category_profit.index, category_profit,
        yerr=category_std, capsize=5, color="lightgreen")
plt.title("Average Profit per Category with Error Bars")
plt.xlabel("Category")
plt.ylabel("Average Profit ($)")
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
```

Q2. Highlight the Maximum Sales Order Using an Annotation.

```python
df_subset = df.head(100)

plt.figure(figsize=(10,5))
plt.plot(df_subset.index, df_subset["Sales"], marker="o", color="blue", label="Sales")

max_idx = df_subset["Sales"].idxmax()
max_val = df_subset["Sales"].max()

plt.annotate("Highest Sales", xy=(max_idx, max_val),
             xytext=(max_idx+5, max_val+50),
             arrowprops=dict(facecolor="red", arrowstyle="->"))

plt.title("Sales Trend with Maximum Highlighted")
plt.xlabel("Order Index")
plt.ylabel("Sales ($)")
plt.legend()
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
```

Q3. Plot Discount per Category with Error Bars and Cap Size.

```python
category_discount = df.groupby("Category")["Discount"].mean()
category_std = df.groupby("Category")["Discount"].std()

plt.figure(figsize=(7,5))
plt.bar(category_discount.index, category_discount, yerr=category_std, capsize=8, color="orange")
plt.title("Average Discount per Category with Error Bars")
plt.xlabel("Category")
plt.ylabel("Discount")
plt.grid(axis="y", linestyle="--", alpha=0.6)
plt.show()
```

Q4. Annotate the Lowest Profit in the First 50 Orders.

```python
df_subset = df.head(50)

plt.figure(figsize=(10,5))
plt.plot(df_subset.index, df_subset["Profit"], marker="s", color="purple", label="Profit")

min_idx = df_subset["Profit"].idxmin()
min_val = df_subset["Profit"].min()

plt.annotate("Lowest Profit", xy=(min_idx, min_val),
             xytext=(min_idx+2, min_val-200),
             arrowprops=dict(facecolor="black", arrowstyle="->"))

plt.title("Profit Trend with Minimum Highlighted")
plt.xlabel("Order Index")
plt.ylabel("Profit ($)")
plt.legend()
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
```

Q5. Combine Error Bars and Annotation: Average Quantity per Category with Highest Quantity Highlighted.

```python
category_quantity = df.groupby("Category")["Quantity"].mean()
category_std = df.groupby("Category")["Quantity"].std()

plt.figure(figsize=(8,5))
plt.bar(category_quantity.index, category_quantity, yerr=category_std, capsize=5, color="skyblue")

max_idx = category_quantity.idxmax()
max_val = category_quantity.max()

plt.annotate("Highest Avg Quantity", xy=(max_idx, max_val),
             xytext=(max_idx, max_val+2),
             arrowprops=dict(facecolor="red", arrowstyle="->"))

plt.title("Average Quantity per Category with Error Bars & Annotation")
plt.xlabel("Category")
plt.ylabel("Quantity")
plt.grid(axis="y", linestyle="--", alpha=0.6)
plt.show()
```

### Summary

Error bars and annotations are essential tools in data visualization that add **clarity, precision, and storytelling power** to plots. Error bars communicate **uncertainty or variability** in the data. For example, when plotting the average sales across categories in the Superstore dataset, error bars derived from standard deviation show how much the values differ from the mean. A small error bar suggests stable, consistent sales, while a large one highlights higher fluctuations. This helps decision-makers identify stable vs. risky product categories.

Annotations, on the other hand, are about **directing attention**. In a dataset with hundreds of points, an important outlier or trend can easily be overlooked. By annotating specific points, such as the **highest profit order**, we immediately guide the audience to the most relevant insight. This technique transforms a simple chart into a **narrative tool**, making it easier to explain findings in reports or presentations.

Together, error bars and annotations bring balance: error bars provide **statistical context** (showing variability), while annotations add **storytelling context** (highlighting significance). In AI/ML applications, error bars can represent prediction confidence intervals, while annotations can highlight misclassified points or key decision boundaries.

In short, these two features elevate visualizations from being purely descriptive to being **insight-driven and communicative**, making them highly effective in business analytics and research presentations.
