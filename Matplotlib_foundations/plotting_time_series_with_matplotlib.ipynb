# Plotting Time Series with Matplotlib

### What is Time Series Plotting?

A **time series** is a sequence of data points collected or recorded at specific intervals of time (e.g., daily sales, monthly temperature, stock prices, sensor data). Time series plotting is the process of visualizing this data where the **X-axis represents time** and the **Y-axis represents the observed values**.

Unlike ordinary plots, time series charts emphasize **trends, seasonality, and patterns** across time. Matplotlib integrates well with **datetime objects**, making it easier to create clear visualizations. It allows for formatting date ticks, rotating labels, and applying moving averages or rolling windows to highlight trends.

In **AI/ML**, time series visualization is vital in:

- Forecasting stock prices or demand.
- Tracking loss/accuracy during model training.
- Monitoring sensor data in IoT or anomaly detection.
- Visualizing sequences in NLP or reinforcement learning.

Thus, mastering time series plotting is essential for analyzing temporal data and identifying insights hidden in patterns over time.

**1. Basic Time Series Plot**

```python
import matplotlib.pyplot as plt
import pandas as pd

# Sample time series data
dates = pd.date_range("2023-01-01", periods=10)
values = [100, 120, 115, 130, 150, 160, 155, 170, 165, 180]

plt.figure(figsize=(10,5))
plt.plot(dates, values, marker="o", linestyle="-", color="blue")
plt.title("Basic Time Series Plot")
plt.xlabel("Date")
plt.ylabel("Value")
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
```

**2. Time Series with Multiple Lines**

```python
dates = pd.date_range("2023-01-01", periods=10)
sales = [200, 210, 220, 230, 250, 240, 260, 280, 300, 310]
profit = [50, 55, 53, 60, 62, 65, 70, 75, 80, 90]

plt.figure(figsize=(10,5))
plt.plot(dates, sales, label="Sales", color="blue")
plt.plot(dates, profit, label="Profit", color="green")
plt.title("Sales and Profit Over Time")
plt.xlabel("Date")
plt.ylabel("Value")
plt.legend()
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
```

**3. Time Series with Rolling Average**

```python
import numpy as np

# Simulate noisy data
dates = pd.date_range("2023-01-01", periods=30)
values = np.random.randint(50, 100, size=30)

# Rolling average
rolling_mean = pd.Series(values).rolling(window=5).mean()

plt.figure(figsize=(12,5))
plt.plot(dates, values, label="Original Data", alpha=0.7)
plt.plot(dates, rolling_mean, label="5-Day Rolling Average", linewidth=2, color="red")
plt.title("Time Series with Rolling Average")
plt.xlabel("Date")
plt.ylabel("Value")
plt.legend()
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
```

**4. Time Series with Custom Date Formatting**

```python
import matplotlib.dates as mdates

dates = pd.date_range("2023-01-01", periods=12, freq="M")
values = [100, 120, 115, 140, 160, 170, 180, 175, 190, 200, 210, 220]

plt.figure(figsize=(12,5))
plt.plot(dates, values, marker="o")
plt.title("Monthly Data with Custom Date Formatting")
plt.xlabel("Month")
plt.ylabel("Value")

# Format x-axis for month/year
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter("%b %Y"))
plt.gca().xaxis.set_major_locator(mdates.MonthLocator(interval=2))
plt.xticks(rotation=45)
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
```

### Exercises

Q1. Plot daily Sales data for the first 30 days using a line plot.

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Sample daily sales data
dates = pd.date_range("2023-01-01", periods=30, freq="D")
sales = np.random.randint(100, 300, size=30)

plt.figure(figsize=(10,5))
plt.plot(dates, sales, marker="o", color="blue", linestyle="-")
plt.title("Daily Sales for First 30 Days")
plt.xlabel("Date")
plt.ylabel("Sales ($)")
plt.xticks(rotation=45)
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
```

Q2. Create a time series with both Sales and Profit, using different line styles.

```python
dates = pd.date_range("2023-01-01", periods=15, freq="D")
sales = np.random.randint(200, 500, size=15)
profit = np.random.randint(50, 150, size=15)

plt.figure(figsize=(10,5))
plt.plot(dates, sales, label="Sales", linestyle="-", color="blue")
plt.plot(dates, profit, label="Profit", linestyle="--", color="green")
plt.title("Sales and Profit Time Series")
plt.xlabel("Date")
plt.ylabel("Value")
plt.legend()
plt.xticks(rotation=45)
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
```

Q3. Plot a rolling mean (7-day window) of random temperature data.

```python
dates = pd.date_range("2023-01-01", periods=30, freq="D")
temperature = np.random.randint(15, 35, size=30)  # random daily temperatures

# Compute rolling mean
rolling_mean = pd.Series(temperature).rolling(window=7).mean()

plt.figure(figsize=(12,5))
plt.plot(dates, temperature, label="Daily Temperature", alpha=0.6, color="orange")
plt.plot(dates, rolling_mean, label="7-Day Rolling Mean", color="red", linewidth=2)
plt.title("Temperature with 7-Day Rolling Average")
plt.xlabel("Date")
plt.ylabel("Temperature (C)")
plt.legend()
plt.grid(True, linestyle="--", alpha=0.6)
plt.xticks(rotation=45)
plt.show()
```

Q4. Display monthly cumulative sales with custom x-axis labels (e.g., Jan 2023).

```python
import matplotlib.dates as mdates

dates = pd.date_range("2023-01-01", periods=12, freq="M")
monthly_sales = np.random.randint(1000, 5000, size=12)
cumulative_sales = monthly_sales.cumsum()

plt.figure(figsize=(12,5))
plt.plot(dates, cumulative_sales, marker="o", color="purple")
plt.title("Cumulative Monthly Sales")
plt.xlabel("Month")
plt.ylabel("Cumulative Sales ($)")

# Format x-axis labels
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter("%b %Y"))
plt.xticks(rotation=45)
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
```

Q5. Create a dual-axis time series where one axis shows Orders and another shows Returns.

```python
dates = pd.date_range("2023-01-01", periods=20, freq="D")
orders = np.random.randint(50, 200, size=20)
returns = np.random.randint(5, 40, size=20)

fig, ax1 = plt.subplots(figsize=(12,5))

# Plot Orders on primary Y-axis
ax1.plot(dates, orders, color="blue", label="Orders")
ax1.set_xlabel("Date")
ax1.set_ylabel("Orders", color="blue")
ax1.tick_params(axis="y", labelcolor="blue")

# Create secondary Y-axis for Returns
ax2 = ax1.twinx()
ax2.plot(dates, returns, color="red", linestyle="--", label="Returns")
ax2.set_ylabel("Returns", color="red")
ax2.tick_params(axis="y", labelcolor="red")

plt.title("Orders vs Returns Time Series")
fig.tight_layout()
plt.show()
```

This makes it easier to analyze both short-term fluctuations and long-term trends.

- Businesses use it to track **sales and profits**.
- Stock analysts monitor **prices and volumes**.
- Machine learning practitioners visualize **training loss and accuracy over epochs**.
- Researchers analyze **climate, population, or sensor data**.

By mastering time series visualization, we gain the ability to interpret **temporal relationships**, spot patterns, and make data-driven forecasts, which are essential in AI, finance, operations, and research.
