# Exporting Publication-Quality Figures

### What is Exporting Publication-Quality Figures?

When working with **data visualization**, it’s not enough to simply create a plot — we often need to **export high-quality figures** for use in **reports, presentations, papers, or publications**. A **publication-quality figure** is one that is **visually clear, properly labeled, and high-resolution**, ensuring it looks professional in print or on screen.

In **Matplotlib**, exporting figures means **saving plots in specific formats (like PNG, PDF, SVG, EPS)** with control over **size, resolution (DPI), color space, and font quality**. These settings are essential for making visuals that meet the standards of **academic journals, research reports, and corporate presentations**.

High-quality figures make our findings **credible, reproducible, and visually appealing**. They also ensure consistency across publications — especially in **AI/ML research**, where figures often display results, loss curves, or model comparisons.

### Why is this Important?

- **Clarity & Readability:** Journals and reports demand figures that remain clear even after scaling or printing.
- **Professionalism:** High-quality exports improve presentation quality and trustworthiness.
- **Versatility:** Supports multiple file formats for web, print, and vector graphics.
- **Reproducibility:** Same figure can be regenerated programmatically with consistent results.

### Common Export Formats

| Format | Description | Best Use |
| --- | --- | --- |
| **PNG** | Raster image (pixels), supports transparency | Web reports, presentations |
| **PDF** | Vector format (scalable) | Research papers, printing |
| **SVG** | XML-based vector graphics | Web-based dashboards |
| **EPS** | Encapsulated PostScript | Academic publishing, LaTeX |
| **JPG** | Compressed raster | Basic image sharing |

**Examples**

1. Basic Figure Saving

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y, label="Sine Wave", color="blue", linewidth=2)
plt.title("Basic Sine Wave")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.legend()

# Save figure
plt.savefig("save_plots/sine_wave.png", dpi=300)
plt.show()
```

2. Exporting with Different Formats and DPI

```python
plt.plot(x, y, color="red", label="Sine Wave")
plt.title("High-Quality Export Example")
plt.legend()

# Export in multiple formats
plt.savefig("sine_wave_highres.png", dpi=600)   # high-resolution PNG
plt.savefig("sine_wave_vector.pdf")              # vector PDF for publication
plt.savefig("save_plots/sine_wave.svg")                     # scalable SVG for web
plt.show()
```

3. Transparent Background & Tight Layout

```python
plt.plot(x, y, color="purple", linewidth=3)
plt.title("Transparent Figure Example")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")

# Export with transparency and proper padding
plt.savefig("save_plots/transparent_plot.png", dpi=300, transparent=True, bbox_inches="tight")
plt.show()
```

4. Controlling Figure Size and Font for Publication

```python
plt.figure(figsize=(8, 5))  # inches
plt.plot(x, np.cos(x), color="green", linewidth=2)
plt.title("Customized Figure", fontsize=16)
plt.xlabel("X", fontsize=12)
plt.ylabel("cos(X)", fontsize=12)

plt.savefig("save_plots/customized_figure.pdf", dpi=300, bbox_inches="tight")
plt.show()
```

5. Using rcParams for Consistent Style

```python
plt.rcParams["figure.dpi"] = 200
plt.rcParams["savefig.dpi"] = 300
plt.rcParams["font.size"] = 12
plt.rcParams["axes.labelweight"] = "bold"

plt.plot(x, y, color="orange", label="Styled Plot")
plt.legend()
plt.title("Styled Export Example")

plt.savefig("save_plots/styled_export.png", bbox_inches="tight")
plt.show()
```

### Exercises

Q1. Create a plot of Sales vs Profit and export it as a high-resolution PNG (300 DPI).

```python
import pandas as pd
df = pd.read_csv("data/Superstore.csv", encoding='latin-1')

plt.scatter(df["Sales"], df["Profit"], alpha=0.6, color="teal")
plt.title("Sales vs Profit")
plt.xlabel("Sales ($)")
plt.ylabel("Profit ($)")
plt.savefig("save_plots/sales_profit_highres.png", dpi=300)
plt.show()
```

Q2. Export the same plot as a vector-based PDF.

```python
plt.scatter(df["Sales"], df["Profit"], alpha=0.6, color="orange")
plt.title("Sales vs Profit (Vector PDF)")
plt.xlabel("Sales ($)")
plt.ylabel("Profit ($)")
plt.savefig("save_plots/sales_profit_vector.pdf", bbox_inches="tight")
plt.show()
```

Q3. Generate a figure with transparent background and save as `SVG`.

```python
plt.plot(df["Sales"].head(50), df["Profit"].head(50), color="purple")
plt.title("Transparent Export Example")
plt.savefig("save_plots/transparent_figure.svg", transparent=True, bbox_inches="tight")
plt.show()
```

Q4. Use `rcParams` to set default figure DPI and font size, then save a plot.

```python
plt.rcParams["savefig.dpi"] = 400
plt.rcParams["font.size"] = 13
plt.plot(df["Quantity"].head(50), color="green")
plt.title("Styled Export")
plt.savefig("save_plots/styled_plot.png", bbox_inches="tight")
plt.show()
```

Q5. Export multiple subplots in one figure as a publication-ready PDF using `tight_layout()`.

```python
fig, axs = plt.subplots(1, 2, figsize=(10, 4))
axs[0].hist(df["Sales"], bins=30, color="blue", alpha=0.7)
axs[0].set_title("Sales Distribution")
axs[1].hist(df["Profit"], bins=30, color="red", alpha=0.7)
axs[1].set_title("Profit Distribution")
plt.tight_layout()
plt.savefig("save_plots/publication_ready.pdf", dpi=300, bbox_inches="tight")
plt.show()
```

### Summary

Exporting publication-quality figures ensures that our visualizations are **clear, professional, and suitable for academic or business use**. By controlling parameters like **DPI (dots per inch)**, **figure size**, **font style**, and **file format**, we can create visuals that retain their clarity and detail across all mediums — whether in printed journals, digital reports, or online dashboards.

Matplotlib’s `savefig()` function allows precise customization: we can export in multiple formats (like **PNG, PDF, SVG, EPS**), set **transparent backgrounds**, and use **tight layouts** for perfect spacing. The use of **vector formats (PDF, SVG)** is crucial in research because they scale infinitely without losing quality.

In data science and AI/ML workflows, such exports are used in **papers, dashboards, and presentations** to communicate insights effectively. Ultimately, mastering figure export transforms our visualizations from simple charts into **publication-grade graphics** that are both **technically accurate and visually impactful**.
