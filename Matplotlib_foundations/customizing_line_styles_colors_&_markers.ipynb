# Customizing Line Styles, Colors, and Markers

### What is Line Customization in Matplotlib?

In Matplotlib, a **line plot** is one of the most common ways to visualize data trends. However, the default line style may not always be clear, visually appealing, or suitable for presentation. Customizing **line styles, colors, and markers** allows us to **differentiate multiple data series**, **highlight trends**, and make plots **professional and presentation-ready**.

Matplotlib provides several parameters in `plt.plot()` or the **OO interface** to control **line appearance**:

- **Color**: Changes the line color to make it stand out or to indicate categories.
- **Line style**: Controls the appearance of the line (solid, dashed, dotted, dash-dot).
- **Marker**: Adds symbols at each data point to emphasize values or differentiate series.

### Customizing Colors

Colors can be specified in multiple ways:

- **Named colors**: `'red'`, `'blue'`, `'green'`
- **Short codes**: `'r'`, `'b'`, `'g'`
- **Hex codes**: `'#FF5733'`
- **RGBA tuples** for transparency: `(0.1, 0.2, 0.5, 0.7)`

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.plot(x, y, color='purple')  # Named color
plt.plot(x, y, color='#FF5733') # Hex code
```

### Customizing Line Styles

Common **line styles**:

- `'-'` → Solid (default)
- `'--'` → Dashed
- `':'` → Dotted
- `'-.'` → Dash-dot

```python
plt.plot(x, y, linestyle='--')  # Dashed line
```

### Customizing Markers

Markers highlight individual data points:

- `'o'` → Circle
- `'s'` → Square
- `'^'` → Triangle
- `'x'` → Cross

```python
plt.plot(x, y, marker='o', markersize=8, markerfacecolor='red')
```

Other marker parameters include **marker size** (`markersize`) and **marker color** (`markerfacecolor` / `markeredgecolor`).

### Example with Superstore Dataset

```python
import pandas as pd

df = pd.read_csv("data/Superstore.csv", encoding='latin-1')
sales = df['Sales'][:50]
profit = df['Profit'][:50]

plt.figure(figsize=(10,5))
plt.plot(sales, color='blue', linestyle='-', marker='o', label='Sales')
plt.plot(profit, color='red', linestyle='--', marker='x', label='Profit')
plt.title("Sales vs Profit with Custom Styles")
plt.xlabel("Order Index")
plt.ylabel("Value")
plt.legend()
plt.grid(True)
plt.show()
```

This plot clearly differentiates **Sales and Profit** using color, line style, and markers.

### Why It Matters in AI/ML and Data Science

- Improves **readability** and **visual appeal** of plots.
- Makes **multiple data series distinguishable** in one figure.
- Useful for **presentations, reports, and dashboards**.
- Helps in AI/ML to **compare training/validation curves**, feature importance trends, or predictions vs actual values.

### Summary

Customizing **line styles, colors, and markers** in Matplotlib is an essential technique to make plots **clear, informative, and visually appealing**. By adjusting these properties, multiple data series can be easily distinguished in a single figure, improving readability and helping convey insights more effectively.

**Colors** allow visual separation and can be specified using names, short codes, hex codes, or RGBA values for transparency. **Line styles**—such as solid, dashed, dotted, or dash-dot—help differentiate trends or categories, while **markers** highlight individual data points and make patterns easier to see. Additional marker properties like size and edge/face color provide further customization.
