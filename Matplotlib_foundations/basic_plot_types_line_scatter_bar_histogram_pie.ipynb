# Basic Plot Types in Matplotlib (Line, Scatter, Bar, Histogram, Pie)

When we start visualizing data, the first step is learning the **basic plot types**. Matplotlib provides flexible tools to create different kinds of plots that help us **understand distributions, relationships, and comparisons** in our dataset.

### Line Plot

A **line plot** is used to show trends over an index (time, order, etc.). It connects data points with straight lines.

- Good for showing continuous data or trends.

```python
import matplotlib.pyplot as plt
import pandas as pd

# Load dataset
df = pd.read_csv("data/Superstore.csv", encoding='latin-1')

plt.plot(df["Sales"].head(50))
plt.title("Line Plot of Sales")
plt.xlabel("Order Index")
plt.ylabel("Sales")
plt.show()
```

### Scatter plot

A **scatter plot** shows the relationship between two numerical variables.

- Each point represents an observation.
- Good for correlation and distribution analysis.

```python
plt.scatter(df["Sales"].head(50), df["Profit"].head(50), color="red")
plt.title("Scatter Plot: Sales vs Profit")
plt.xlabel("Sales")
plt.ylabel("Profit")
plt.show()
```

### Bar plot

A **bar plot** compares quantities across categories.

- In our dataset, we can compare average sales per category.

```python
category_sales = df.groupby("Category")["Sales"].sum()
category_sales.plot(kind="bar", color="skyblue")
plt.title("Bar Plot of Sales by Category")
plt.xlabel("Category")
plt.ylabel("Total Sales")
plt.show()
```

### Histrogram

A **histogram** shows the distribution of a single variable by grouping values into **bins**.

- Useful to see the spread of data (like sales distribution).

```python
plt.hist(df["Sales"], bins=20, color="green", edgecolor="black")
plt.title("Histogram of Sales")
plt.xlabel("Sales")
plt.ylabel("Frequency")
plt.show()
```

### Pie Chart

A **pie chart** shows proportions of a whole.

- Best for visualizing percentage breakdown of categories.

```python
category_sales = df.groupby("Category")["Sales"].sum()
category_sales.plot(kind="pie", autopct='%1.1f%%', startangle=90, colormap="Set3")
plt.title("Sales Distribution by Category")
plt.ylabel("")  # remove default y-label
plt.show()
```

### Exercises

Q1. Plot Sales for the first 100 rows using a line plot with proper labels and title.

```python
import matplotlib.pyplot as plt
import pandas as pd

# Load dataset
df = pd.read_csv("data/Superstore.csv", encoding='latin-1')

# Line Plot
plt.figure(figsize=(8,5))
plt.plot(df["Sales"].head(100), color="blue")
plt.title("Sales Trend (First 100 Orders)")
plt.xlabel("Order Index")
plt.ylabel("Sales ($)")
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
```

Q2. Create a scatter plot of Profit vs Sales for the first 75 rows.

```python
plt.figure(figsize=(8,5))
plt.scatter(df["Sales"].head(75), df["Profit"].head(75), color="red", alpha=0.7)
plt.title("Scatter Plot: Profit vs Sales (First 75 Orders)")
plt.xlabel("Sales ($)")
plt.ylabel("Profit ($)")
plt.grid(True, linestyle=":")
plt.show()
```

Q3. Make a bar chart of total Quantity per Category.

```python
category_quantity = df.groupby("Category")["Quantity"].sum()

plt.figure(figsize=(7,5))
category_quantity.plot(kind="bar", color="orange", edgecolor="black")
plt.title("Total Quantity per Category")
plt.xlabel("Category")
plt.ylabel("Total Quantity")
plt.grid(axis="y", linestyle="--", alpha=0.6)
plt.show()
```

Q4. Plot a histogram of Profit with 30 bins.

```python
plt.figure(figsize=(8,5))
plt.hist(df["Profit"], bins=30, color="green", edgecolor="black", alpha=0.7)
plt.title("Profit Distribution (30 bins)")
plt.xlabel("Profit ($)")
plt.ylabel("Frequency")
plt.grid(axis="y", linestyle="--", alpha=0.6)
plt.show()
```

Q5. Create a pie chart of Discount by taking the mean Discount per Category.

```python
mean_discount = df.groupby("Category")["Discount"].mean()

plt.figure(figsize=(6,6))
mean_discount.plot(kind="pie", autopct='%1.1f%%', startangle=90, colormap="Set3")
plt.title("Average Discount by Category")
plt.ylabel("")  # remove default y-axis label
plt.show()
```

### Summary

Matplotlib supports different **basic plot types**:

- **Line plots**  best for trends and continuous data.
- **Scatter plots**  show relationships and correlations between two variables.
- **Bar plots**  compare quantities across categories.
- **Histograms**  show distributions and spread of values.
- **Pie charts**  display proportions of a whole.

Each plot serves a different analytical purpose. For example, **line plots** help track sales over time, **scatter plots** reveal whether sales and profit are correlated, **bar charts** summarize categories, **histograms** show skewness/outliers in sales, and **pie charts** quickly show proportions like category share.

By mastering these, well have the essential toolkit for **exploratory data analysis (EDA)** and communicating insights effectively.
