# Plotting with Pyplot vs Object-Oriented Interface

### What is the Pyplot API?

The **Pyplot API** (`matplotlib.pyplot`) is the **simplest and most commonly used interface** in Matplotlib. It is designed to mimic MATLABâ€™s style, where plots are created by calling functions directly. Each function works on the **current figure and axes** without requiring explicit object management.

### Pyplot Interface

The **Pyplot API** (`plt`) is a **state-machine style interface**, which means it keeps track of the current Figure and Axes automatically. This is convenient for **quick and simple plots**. You can create a figure, plot your data, and add labels and titles without explicitly managing Figures or Axes objects.

**Example**

```python
import matplotlib.pyplot as plt

plt.plot([1,2,3,4], [10,20,25,30])
plt.title("Simple Pyplot Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

Here, Matplotlib automatically creates a **Figure** and an **Axes** object behind the scenes. Pyplot is great for **quick, simple, and interactive plotting**, but it becomes limited when building **complex visualizations with multiple subplots**.

### What is the Object-Oriented (OO) Interface?

The **Object-Oriented API** gives **direct control over Figures and Axes objects**. Instead of working implicitly, you explicitly create and manage the objects.

```python
fig, ax = plt.subplots(figsize=(6,4))
ax.plot([1,2,3,4], [10,20,25,30], color="red")
ax.set_title("Object-Oriented Example")
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
plt.show()
```

This approach is more **flexible and professional**. It allows customization of **multiple axes, legends, grids, and styles** within the same figure, making it suitable for **dashboards, reports, and publication-ready charts**.

### Pyplot vs OO Interface: Key Differences

- **Pyplot API**
    - Quick and easy for small plots.
    - Implicit figure/axes creation.
    - Example: `plt.plot(x, y)`.
- **OO Interface**
    - Explicit, more control.
    - Essential for complex plots with subplots/twin axes.
    - Example: `fig, ax = plt.subplots(); ax.plot(x, y)`.

### Why it matters in AI/ML and Data Science

- Pyplot is great for **quick EDA** and **simple plots**.
- OO interface is ideal for **dashboards, reports, and multi-panel visualizations**.
- Both are important for **presentation-ready visualizations**, trend analysis, and monitoring ML models.

### Example with Superstore Dataset

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("data/Superstore.csv", encoding='latin-1')

# Pyplot Example
plt.plot(df['Sales'][:50], label="Sales")
plt.title("Sales using Pyplot")
plt.xlabel("Index")
plt.ylabel("Sales")
plt.legend()
plt.show()

# OO Example
fig, ax = plt.subplots(figsize=(8,5))
ax.plot(df['Profit'][:50], color="green", label="Profit")
ax.set_title("Profit using OO Interface")
ax.set_xlabel("Index")
ax.set_ylabel("Profit")
ax.legend()
plt.show()
```

### Exercises

Q1. Create a blank Figure object of size (6,4).

```python
fig = plt.figure(figsize=(6,4))
plt.show()
```

Q2. Add an Axes to the Figure manually using `fig.add_axes([0.1,0.1,0.8,0.8])` and plot some random data.

```python
import numpy as np

fig = plt.figure(figsize=(6,4))
ax = fig.add_axes([0.1,0.1,0.8,0.8])
x = np.arange(10)
y = np.random.randint(5,20,10)
ax.plot(x, y, marker='o', color='purple')
ax.set_title("Manual Axes Plot with Random Data")
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
plt.show()
```

Q3. Use `plt.subplots(2,2)` to create a 2x2 grid of plots and fill each subplot with different charts (line, bar, scatter, histogram).

```python
fig, axes = plt.subplots(2,2, figsize=(10,8))

axes[0,0].plot(df['Sales'][:50], color='blue')
axes[0,0].set_title("Line Plot of Sales")

axes[0,1].bar(np.arange(20), df['Profit'][:20], color='orange')
axes[0,1].set_title("Bar Plot of Profit")

axes[1,0].scatter(df['Sales'][:50], df['Profit'][:50], color='green')
axes[1,0].set_title("Scatter: Sales vs Profit")

axes[1,1].hist(df['Sales'], bins=20, color='purple', alpha=0.7)
axes[1,1].set_title("Histogram of Sales")

plt.tight_layout()
plt.show()
```

Q4. Customize titles, axis labels, and grids for each subplot.

```python
fig, axes = plt.subplots(2,2, figsize=(10,8))

axes[0,0].plot(df['Sales'][:50], color='blue')
axes[0,0].set_title("Line Plot of Sales")
axes[0,0].set_xlabel("Index")
axes[0,0].set_ylabel("Sales")
axes[0,0].grid(True)

axes[0,1].bar(np.arange(20), df['Profit'][:20], color='orange')
axes[0,1].set_title("Bar Plot of Profit")
axes[0,1].set_xlabel("Index")
axes[0,1].set_ylabel("Profit")
axes[0,1].grid(True)

axes[1,0].scatter(df['Sales'][:50], df['Profit'][:50], color='green')
axes[1,0].set_title("Scatter: Sales vs Profit")
axes[1,0].set_xlabel("Sales")
axes[1,0].set_ylabel("Profit")
axes[1,0].grid(True)

axes[1,1].hist(df['Sales'], bins=20, color='purple', alpha=0.7)
axes[1,1].set_title("Histogram of Sales")
axes[1,1].set_xlabel("Sales")
axes[1,1].set_ylabel("Frequency")
axes[1,1].grid(True)

plt.tight_layout()
plt.show()
```

Q5. Create two Axes in the same Figure with different scales (Sales vs Profit).

```python
fig, ax1 = plt.subplots(figsize=(7,5))

ax1.plot(df['Sales'][:50], color='blue', label='Sales')
ax1.set_xlabel("Index")
ax1.set_ylabel("Sales", color='blue')
ax1.tick_params(axis='y', labelcolor='blue')
ax1.legend(loc="upper left")

ax2 = ax1.twinx()
ax2.plot(df['Profit'][:50], color='red', label='Profit')
ax2.set_ylabel("Profit", color='red')
ax2.tick_params(axis='y', labelcolor='red')
ax2.legend(loc="upper right")

plt.title("Sales vs Profit with Twin Axes")
fig.tight_layout()
plt.show()
```
